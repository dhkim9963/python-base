# 클래스와 매직 메서드
파이썬에는 클래스가 존재함\
프로그래밍 언어에서 클래스가 존재한다는 것은 사용자-정의 타입(user-defined type)을 작성하여 능력을 부여할 수 있다는 것을 의미함\
클래스를 정의하려면 어떤 데이터를 담고 무엇을 할 수 있는지 명시해야 하는데, 현대 프로그래밍 언어 대부분은 이 기능을 가지고 있음\
파이썬은 클래스에 독특한 기능을 더했는데, 특정 상황에서 자동으로 호출되어 응답하는 매직 메서드가 바로 그것임

## 1. 클래스와 객체 기본 문법
파이썬에서 클래스를 정의하는 기본 문법은 다음과 같음
```
class 클래스_이름:
    문장
```
'문장'은 동일한 들여쓰기(indent)를 1개 이상의 줄로 작성된 코드로 구성됨\
문장에는 내용이 반드시 있어야 하며, 연산 처리할 것이 없다(no-op)는 의미로 pass 키워드를 사용할 수 있음\
이 pass 키워드는 추후에 클래스를 정의할 때 유용하게 사용 가능함\
예를 들어 다음과 같이 Car 클래스를 정의할 수 있음

````python
class Car:
    pass
````

파이썬에서 '인스턴스'라고 부르는 클래스 실체를 원하는 만큼 생성할 수 있음\
다음 코드는 Car 클래스 인스턴스를 여러 개 생성하는 예시임

````python
car1 = Car()
car2 = Car()
car3 = Car()
````

클래스에 변수를 생성하면 이 변수들은 '클래스 변수'가 될 것이며, 모든 객체에 공유됨\
예를 들어 Car 클래스를 다음과 같이 정의했다고 가정해보자

````python
class Car:
    accel = 3.0
    mpg = 25
````

이제 Car 클래스 인스턴스를 출력하면 이 변수들의 값을 생성할 것임

````python
print('car1.accel = ', car1.accel)
print('car2.accel = ', car2.accel)
print('car1.mpg = ', car1.mpg)
print('car2.mpg = ', car2.mpg)
````

이 코드에는 독특한 기능이 숨겨져 있는데, Car 인스턴스는 ㅁccel 변수에 해당 인스턴스만을 위한 값을 대입할 수 있음\
이는 3.0이었던 클래스 변수 값을 교체(override)함\
다음과 같이 my_car 인스턴스를 생성하여 accel 변수에 값을 대입해 보자

````python
my_car = Car()
yr_car = Car()
my_car.accel = 5.0

print(my_car.accel)
print(yr_car.accel)
````

## 2. 인스턴스 변수에 대해 더 알아보기
다른 프로그래밍 언어와는 다르게 파이썬의 인스턴스 변수는 클래스 안에서 직접 생성되지는 않음\
필요할 때마다 바로 생성되거나(ad hoc) __init__ 메서드 안에서 생성됨\
인스턴스 변수를 생성하는 일반적인 규칙은 다른 변수를 생성하는 것과 같음\
즉, 값을 대입한느 것이며 인스턴스 변수에 값을 대입하기 위해서는 점 기호(.)문법을 사용해야 함
```
객체.변수_이름 = 값
```
예를 들어 정의한 Dog 클래스로 생성한 인스턴스에 여러 속성(attribute)을 추가하면, 이 속성들이 바로 인스턴스 변수가 됨

````python
class Dog:
    pass
````

````python
my_dog = Dog()
my_dog.name = 'Champ the Wonder Dog'
my_dog.breed = 'Great Dane'
my_dog.age = 5
````

이제 데이터 변수 3개가 my_dog 객체에 할당되었으며 name, breed, age는 다음과 같이 접근할 수 있음

````python
print('Breed and age are {} and {}.'.format(my_dog.breed, my_dog.age))
````

이 시점에 name, breed, age는 my_dog 객체만을 위한 속성임\
다른 Dog 객체들은 동일한 속성을 가질 필요가 없음(이것이 바로 인스턴스 변수)\
하지만 필요한 시점에 언제든지 동일한 속성을 다른 객체에 추가할 수 있음

````python
yr_dog = Dog()
top_dog = Dog()
hot_dog = Dog()
hot_dog.name = 'Hotty Totty'
hot_dog.breed = 'Dachshund'
````

어떻게 하면 동일한 클래스의 모든 객체가 같은 속성들(인스턴스 변수)을 가질 수 있을까?\
정답은 바로 __init__ 메서드를 활용하는 것

## 3. __init__메서드와 __new__메서드
__init__메서드를 정의한 모든 클래스는 객체가 생성될 때 자동으로 __init__ 메서드가 호출(invoke)됨\
클래스의 모든 인스턴스가 동일한 공통 변수들의 집합을 가지면서 각 인스턴스별로 독립적인 값을 지니게 할 때 사용할 수 있음\
```python
class 클래스_이름:
    def __init__(self, args):
        문장
```
self라는 단어는 반드시 사용해야 하는 키워드가 아니며, 생성된 객체를 참조하기 위해 첫 번째 인수로 배치됨\
이 인수는 규칙에 맞는 어떤 이름이라도 상관없지만, 보편적으로 self를 사용함

args는 인수가 여러 개일 수도 있다는 것을 뜻하며(여러 개일 때는 쉼표로 구분), 객체를 생성할 때 넘겨줌\
예를들어 Dog 클래스의 정의문을 __init__메서드를 포함하게 변경할 수 있음

````python
class Dog:
    def __init__(self, name, breed, age):
        self.name = name
        self.breed = breed
        self.age = age
````

이제 클래스 Dog의 객체를 생성할 때 반드시 인수를 3개 입력해야 하며, 이 인수들은 __init__메서드로 전달됨

````python
top_dog = Dog('Handsome Dan', 'Bulldog', 10)
````

앞 코드는 top_dog 객체를 생성했으며, __init__메서드를 호출하였음\
이 경우 __init__메서드는 다음 코드를 실행한 것과 같이 인스턴스 변수들에 값을 대입함
```python
top_dog.name = 'Handsome Dan'
top_dog.breed = 'Bulldog'
top_dog.age = 10
```
이와 유사하게 good_dog 객체를 생성하면서 상위 클래스의 함수에 다른 데이터를 넘겨줄 수도 있음\
```python
good_dog = Dog('WonderBoy', 'Collie', 11)
```
__init__메서드는 다른 초기화 작업도 할 수 있지만 보통 다음과 같은 패턴으로 작성함
```python
class 클래스_이름:
    def __init__(self, 인수1, 인수2, ...):
        self.인스턴스_변수1 = 인수1
        self.인스턴스_변수2 = 인수2
        ...
```
파이썬은 사실 __new__메서드로 객체를 생성하지만 대부분 초기화 작업은 __init__메서드에서 수행됨\
단 다음 두가지 예외 상황을 제외함
* 메모리 할당을 위한 특별한 기법을 사용하는 경우
* 하위 클래스가 불변(immutable)이거나 내장(built-in)클래스인 경우

## 4. 클래스와 선행 참조 문제
파이썬 함수가 선행 참조(forward reference)이슈가 있듯이, 파이썬 클래스도 같은 이슈가 존재함\
클래스가 반드시 인스턴스화되기 전에 정의되어야 한다는 뜻\
클래스를 인스턴스화한다는 것은 객체를 생성한다는 뜻\
다음 코드는 클래스의 선행 참조 문제를 보여줌

````python
class Marriage:
    def __init__(self):
        self.wife = Person('f')
        self.husband = Person('m')

a_marriage = Marriage()

class Person:
    def __init__(self, gender):
        self.gender = gender
````

이 간단한 프로그램의 실패 이유를 살펴보자\
클래스 정의문을 포함한 첫 몇 줄은 실행되면서 클래스 Marriage를 유효한 클래스로 만듬\
하지만 여섯 번째 줄에서 이 클래스를 인스턴스화(실제 객체 a_marriage 생성 시도)할 때 문제가 생김\
객체가 생성되면서 __init__메서드가 호출되고, Person클래스의 객체인 wife와 husband를 생성하는 부분이 문제임\
Person 클래스는 아직 정의되지 않았고, 새로운 객체를 생성할 수 없음

해결책은 명확한데, 여섯 번째 줄을 코드 끝으로 옮기기만 하면 되며, 이렇게 하면 두 클래스 모두 인스턴스화되기 전에 정의됨

````python
class Marriage:
    def __init__(self):
        self.wife = Person('f')
        self.husband = Person('m')

class Person:
    def __init__(self, gender):
        self.gender = gender

a_marriage = Marriage()
````

일반적으로 클래스를 위한 선행 참조 이슈는 다음 몇 가지 규칙을 지키면 문제가 되지 않음
* 반드시 모든 클래스를 인스턴스화하기 전에 정의해야 함
* 클래스 상호 간에 인스턴스화를 하거나 스스로 인스턴스를 만드는 클래스를 주의
* 클래스가 다른 클래스를 단방향으로 포함하거나 클래스 인스턴스 참조를 지니는 경우는 일반적으로 문제가 되지 않음\
단 상호 동시 참조(mutual dependencies)는 주의가 필요함

## 5. 메서드 기본
메서드는 여러 측면에서 기본적인 함수와 구분됨
1. 메서드는 클래스 정의문 안에서 정의됨 다음과 같은 코드는 메서드를 호출하고 있다는 것을 의미함
```python
my_obj.a_method(12)
```
2. 메서드는 항상 인스턴스를 통해 호출되며, 숨겨진 인수가 전달됨(해당 객체의 참조인 'self'라고 불리는 것임)

다음을 보면 반드시 클래스 정의문 내부에서 정의해야 하는 메서드 정의문과 해당 메서드 호출 방법이 서로 일치하지 않다는 것을 알 수 있음\
```python
class 클래스_이름:
    def 메서드_이름(self, 인수1, 인수2, 인수3, ...):
        문장

객체_이름 = 클래스_이름()
객체_이름.메서드_이름(인수1, 인수2, 인수3, ...)
```
메서드 정의문에서는 숨겨진 첫 번째 인수인 self가 있는 반면, 메서드를 호출할 때는 존재하지 않음\
따라서 함수를 호출할 때보다 정의문에 인수가 하나 더 많은 것임

다음 클래스 정의문 예제를 살펴보자

````python
class Pretty:
    def __init__(self, prefix):
        self.prefix = prefix
        
    
    def print_me(self, a, b, c):
        print(self.prefix, a, sep='')
        print(self.prefix, b, sep='')
        print(self.prefix, c, sep='')
````

Pretty 클래스를 만들었으니, 이제 인스턴스를 생성하고 사용하면서 테스트해 보자\
메서드 호출 시에는 정의할 때 사용한 인수 개수보다 1개 적은 인수를 집어넣고 있다는 것에 주목\
정의문에는 self를 명시적으로 포함하지만 호출할 때는 인수에 넣지 않기 때문

````python
printer = Pretty('-->')
printer.print_me(10, 20, 30)
````

메서드 안에서 메서드가 속한 인스턴스 자체는 항상 self로 참조할 수 있으며, 인스턴스 변수는 'self.변수이름'으로 참조할 수 있음

## 6. 전역 변수/메서드, 지역 변수/메서드
전통적으로 객체 지향 언어를 사용하는 목적 중 하나는 캡슐화\
캡슐화는 클래스 내부 내용이 바깥 세상에 노출되지 못하게 함

파이썬은 스크립팅 언어로 모든 것을 노출하는 경향이 있으며, 모든 것을 볼 수 있어 이 사상과 대치됨\
그래서 파이썬은 다른 언어에 비해 보안성이나 타입 확인 능력이 부족함\
하지만 파이썬은 유용한 규약을 따르고 있는데, 변수와 메서드 이름이 언더스코어 1개로 시작하면 내부용이라는 뜻임\
언더스코어 2개로 시작하는 것은 맹글링(mangling)이라고 불리는 기법으로 의도하지 않은 접근을 통제함\
이 메서드들은 '매직 메서드(magic method)'가 아니라고 가정함

간단한 예제로 이 기법들이 어떻게 동작하는지 확인해 보자\
다음 클래스 정의문에서 변수 x와 변수 y는 외부에서 인스턴스를 통해 접근할 수 있지만, __z는 접근할 수 없음\
반면 클래스 내부에서는 모든 변수 3개(__z 포함)에 접근할 수 있음

````python
class Odd:
    def __init__(self):
        self.x = 10
        self.y = 20
        self.__z = 30
        
    def pr(self):
        print('__z = ', self.__z)
````

주어진 이 클래스 정의문이라면 다음 코드는 완벽하게 유효하며, 기대한 대로 정확하게 동작함

````python
o = Odd()
o.x
````

````python
o.y
````

````python
o.__z
````

마지막 코드에서 에러가 나는 이유는 파이썬이 변수 __z를 클래스 이름과 변수 이름의 조합으로 훼손한(mangled) 이름으로 교체하기 때문\
하지만 __z는 여전히 동일한 클래스의 메서드 정의문 안에서 맹글링 없이 접근이 가능함\
그렇기 때문에 메서드 pr이 여전히 동작하는 것\
변수와 메서드 이름은 동일한 클래스에서 항상 접근할 수 있음\
하지만 파이썬에서는 클래스 내부 참조(intraclass reference)를 보장하기 위해 self를 반드시 사용해야 함

````python
o.pr
````

## 7. 상속
파이썬은 '하위 클래스 만들기(subclassing)'로 상속(inheritance)을 지원함\
작성한 프로그램에서 사용하는 메서드 대부분을 담고 있는 Mammal 클래스가 있다고 가정해 보자\
이런 메서드들 중 일부를 추가하거나 변경할 필요가 생겼음\
가령 Mammal 인스턴스가 할 수 있는 모든 것을 하면서 더 많은 기능을 추가한 Dog 클래스를 만들고 싶다고 해 보자\
하나의 상위(base) 클래스를 단일 상속하는 문법부터 살펴보면
```python
class 클래스_이름(상위_클래스):
    문장
```
이렇게 만들어지 ㄴ클래스는 상위 클래스의 모든 클래스 변수와 메서드들을 상속 받게 됨\
신규 변수나 메서드 정의문을 추가할 수도 있고, 기존 정의문을 재정의할 수도 있음

파이썬의 모든 변수와 메서드 이름은 다형적(polymorphic)이며, 이름은 실행되기 전까지 확정(resolved)되지 않음\
따라서 어떤 객체의 어떤 메서드라도 호출할 수 있으며, 코드가 실행될 때 변수와 메서드 이름이 정확하게 확정될 것임

예를 들어 다음 클래스 구조를 살펴보면 Mammal 상위 클래스와 Dog, Cat 하위 클래스 2개가 존재함\
하위 클래스는 Mammal의 __init__함수와 call_out 메서드를 상속받지만, 각 구현체는 각자의 speak 메서드를 지니고 있음

````python
class Mammal:
    def __init__(self, name, size):
        self.name = name
        self.size = size
    
    def speak(self):
        print('My name is ', name)
    
    def call_out(self):
        self.speak()
        self.speak()
        self.speak()

class Dog(Mammal):
    def speak(self):
        print('ARF!!')
        
class Cat(Mammal):
    def speak(self):
        print('Purrrrrr!!!')
````

````python
my_cat = Cat('Precious', 17.5)
my_cat.call_out()
````

Dog와 Cat 클래스는 __init__ 메서드를 상속받는데 C++의 생성자와는 다름\
그렇다면 하위 클래스가 상위 클래스의 __init__ 정의문을 최대한 활용하면서 추가적인 초기화 작업이 필요하다면 어떻게 해야할까?
해결 방안은 하위 클래스에서 새로운 __init__메서드 정의문을 추가하되, 상위 클래스 버전의 __init__ 메서드를 호출하는 것\
결국 상속을 간접적으로 했어도, 모든 상위 클래스 초기화 메서드를 호출할 수 있게 됨
```python
super().__init__
```
예를 들어 Dog.__init__정의문은 변수 breed를 스스로 초기화하고, super().__init__을 호출하여 초기화 나머지 부분을 수행함

````python
class Dog(Mammal):
    def speak(self):
        print('ARF!')
    def __init__(self, name, size, breed):
        super().__init__(name, size)
        self.breed = breed
````

## 8. 다중 상속
파이썬의 유연한 문법은 다중 상속(multiple inheritance)을 지원함\
이는 클래스를 생성할 때, 2개 이상의 상위 클래스를 상속받을 수 있다는 것
```python
class 클래스_이름(상위_클래스1, 상위_클래스2, ...):
    문장
```
다음 예시에서는 Dog 클래스가 Mammal 클래스뿐만 아니라 다른 클래스 2개도 함께 상속 받고 있음
```python
class Dog(Mammal, Pet, Carnivore):
    def speak(self):
        print('ARF!')
    
    def __init__(self, name, size, breed):
        Mammal.__init__(self, name, size)
        self.breed = breed
```
현재 Dog 클래스는 Mammal 클래스뿐만이 아니라, Pet과 Carnivore 클래스를 상속받게 됨\
따라서 Dog의 모든 인스턴스는(모든 Dog 객체는) 자동으로 Mammal과 같이 Pet과 Carnivore 클래스의 속성을 지니게 됨\
Dog 클래스의 초기화 함수에서는 Mammal.__init__ 메서드만 호출함\
다른 상위 클래스의 초기화 함수는 언제든지 추가될 수 있음\
가령 다음과 같이 가족 별칭(nickname)을 Pet.__init__의 인수로 전달하여 초기화 할 수 있음
```python
def __init__(self, name, size, nickname, breed):
    Mammal.__init__(self, name, size)
    Pet.__init__(self, nickname)
    self.breed = breed
```
다중 상속을 사용하면 충돌이 발생할 수 있음\
가령 3개의 서로 다른 상위 클래스를 상속받는 클래스를 작성하고 있다면 상위 클래스들 중에 같은 메서드나 같은 클래스 변수를 정의하지 않는 이상 문제가 발생하지 않을 것\
하지만 상위 클래스 간에 동일한 메서드나 클래스 변수 이름을 사용한다면 충돌이 발생함

|`__bool__(self)`|불리언 변환 메서드이며, 이 메서드는 모든 bool 함수를 호출하고 나면 항상 호출됨 <br> while 문이나 if 문과 같은 제어문의 조건식에 객체만 등장하는 경우에도 호출됨 <br> 제대로 동작하면 True 혹은 False를 반환함 <br> 객체가 비었거나(empty), None이거나, 0 값을 가지는 경우에만 False를 반환함 <br> 조건이 필요한 모든 상황에 어떤 객체도 사용할 수 있음 <br> `__bool__`메서드가 전혀 구현되어 있지 않다면 True를 반환하는 것이 기본 동작 방식임 <br> 만약 특정 상황에서 False를 반환하고 싶다면 직접 구현해야 함|

다음 예제는 Point 클래스에서 `__str__`과 `__repr__`메서드를 `__init__`과 함께 작성하는 이론적인 방법을 보여줌\
이해를 돕기 위해 `__str__`과 `__repr__` 메서드는 조금 다른 포맷의 문자열을 반환하고 있음

````python
class Point:
    big_prime_1 = 1200556037
    big_prime_2 = 2444555677
    
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y
    
    def __str__(self):
        s = str(self.x) + ', '
        s += str(self.y)
        return s
    
    def __repr__(self):
        s = 'Point(' + str(self.x) + ', '
        s += str(self.y) + ')'
        return s
    
    def __hash__(self):
        n = self.x * big_prime_1
        return (n+self.y) % big_prime_2
    
    def __bool__(self):
        return x and y
````

````python
pt = Point(3,4)
pt
````

````python
print(pt)
````

pt를 직접 호출하면 해당 객체의 대표 표현 방식으로 출력하기 위해 객체 클래스의 `__repr__`메서드를 호출하게 됨\
하지만 pt가 print함수의 인수로 주어지면 print함수는 객체의 표준 문자열 표현 방식으로 출력하기 위해 `__str__`메서드를 호출함\
이번 예제는 이해를 돕기 위해 `__repr__`메서드가 `__str__`메서드보다 더 긴 문자열을 출력하고 있음

### 10.3 비교 메서드
비교 메서드는 클래스 객체를 ==(같다), !=(같지 않다), 부등식(>, <, >=, <=)을 사용하여 서로 비교할 수 있게 해줌\
최소한 객체가 서로 동일하다는 것을 테스트하는 메서드를 직접 구현하는 것이 좋음\
그렇지 않으면 ==의 기본 동작은 is 연산이 될 것이기 때문임(is 연산자는 객체들의 값을 비교하지 않고, 객체의 주소가 같은지 비교하기 때문에 값이 같더라도 False가 반환되는 것을 방지하기 위해 직접 객체 동일 유무를 테스트하는 메서드를 구현하라고 권유하는 것)

파이썬 비교 연산자의 특별한 기능들을 알아보자
* 클래스의 객체들을 정렬하고 싶다면 미만 연산자(<)를 정의\
가령 클래스의 객체들을 컬렉션에 넣고 정렬하려면 동일 클래스의 다른 객체들과 비교하는 `__lt__`메서드를 구현할 수 있음\
이렇게 하면 min, max와 같이 sort 메서드도 사용할 수 있음

* 하지만 컬렉션이 서로 다른 클래스의 객체들을 동시에 가지고 있다면 어떻게 비교할 것인가?\
비교 클래스는 리플렉션 연산자(객체가 연산자의 우측에 있는 경우에 어떻게 동작하는지 정의하는)를 가지고 있지 않지만 대칭(symmetry)이 있음

* 파이썬은 리플렉션 연산자 대신 대칭 규칙을 적용하여 리플렉션이 수행됨\
만약 A > B라면 파이썬은 B < A라고 추론함\
만약 `__lt__`와 `__gt__`의 조합을 구현했더라도 사실상 `__lt__`가 양쪽 방향을 정의하는 것임

* 이 규칙은 모든 메서드를 구현할 필요 없이 자동으로 여러 연산자를 가질 수 있게 함

다음 코드 예제는 파이썬 3.0에서 같은 클래스의 객체들에 적용되는 모든 비교 연산자를 지원하기 위해 구현해야 할 최소 분량의 코드임\
이 클래스 정의문으로 Dog 객체들만 담고 있는 컬렉션은 정렬이 가능해짐

````python
class Dog:
    def __init__(self, n):
        self.n = n
        
    def __eq__(self, other):
        '''==를 구현하면 !=를 자동으로 얻는다.'''
        return self.n == other.n
    
    def __lt__(self, other):
        """<를 구현하면 >를 자동으로 얻는다."""
        return self.n < other.n
    
    def __le__(self, other):
        '''<=를 구현하면 >=를 자동으로 얻는다.'''
        return self.n <= other.n
````

|메서드 문법|설명|
|:--|:--|
|`__eq__(self, other)`| 동일함 테스트, == 연산자 사용 시 호출됨(구현이 안되었다면 파이썬은 is연산자로 동일함 테스트)|
|`__nq__(self, other)`| 동일하지 않음 테스트, != 연산자 사용 시 호출됨|
|`__gt__(self, other)`| 초과 여부 테스트, > 연산자 사용 시 호출됨|
|`__lt__(self, other)`| 미만 여부 테스트, < 연산자 사용 시 호출됨|
|`__ge__(self, other)`|이상 여부 테스트, >= 연산자 사용 시 호출됨|
|`__le__(self, other)`|이하 여부 테스트, <= 연산자 사용 시 호출됨|

어떤 객체라도 서로 정렬할 수 있게 만드는 파이썬 비교 기능의 대칭 규칙이 어떻게 동작하는지, 그리고 어떻게 <가 크고 작음을 동시에 비교할 수 있게 하는지 살펴보자\
```
A > B 이면 B < A 이다.
A < B 이면 B > A 이다.
A >= B 이면 B <= A 이다.
A <= B 이면 B >= A 이다.
A == B 이면 B == A 이다. 추기로 A != B 는 사실이 아니라는 것도 알 수 있다.
```
동일 클래스의 객체끼리만 비교하고 싶다고 가정해 보자\
이 경우는 비교 연산자 절반을 자동으로 확보할 수 있기 때문에 더 적은 코드를 작성할 수 있음\
숫자들을 정렬하고 싶다고 해보자\
문제는 int, float, decimal, fraction 클래스의 코드에 접근할 수 없다는 것임\
Dog 클래스가 있을 때 Dog 객체를 int와 서로 정렬하게 하려면 다음과 같은 비교가 가능해야 함
```
Dog < Dog
Dog < int
int < Dog
```
마지막 줄의 비교는 `__gt__`를 구현하면 비교 가능해짐\
다음 코드에 숫자와 상호 정렬할 수 있는 Dog 클래스 예제가 있음\
또한 인스턴스 변수 d를 초기화하는 방법에 따라 Dog 클래스는 문자열과도 상호 정렬할 수 있음\
'상호 정렬'할 수 있다는 것은 다른 객체들처럼 같은 리스트 안에 객체들을 넣고, 전체를 정렬할 수 있다는 것을 의미함

````python
class Dog:
    def __init__(self, d):
        self.d = d
        
    def __gt__(self, other):
        if type(other) == Dog:
            return self.d > other.d
        else:
            return self.d > other
    
    def __lt__(self, other):
        if type(other) == Dog:
            return self.d < other.d
        else:
            return self.d < other
        
    # __repr__을 정의하는 것은 __str__고 갖게 되는 것임
    def __repr__(self):
        return "Dog(" + str(self.d) + ")"
````

````python
d1, d5, d10 = Dog(1), Dog(5), Dog(10)
a_list = [50, d5, 100, d1, -20, d10, 3]
a_list.sort()
a_list
````

### 10.4 산술 연산자 메서드
|메서드 문법|설명|
|:--|:--|
|`__add__(self, other)`|덧셈. 이 메서드는 클래스의 인스턴스가 증가 연산자(+)의 좌측에 있을 때 호출됨 <br> other 인수는 연산자 우측에 위치한 참조(reference)|
|`__sub__(self, other)`|뺄셈. 이 메서드는 클래스의 인스턴스가 감소 연산자(-)의 좌측에 있을 때 호출됨 <br> other 인수는 연산자 우측에 위치한 참조|
|`__mul__(self, other)`|곱셈. 이 메서드는 클래스의 인스턴스가 곱하기 연산자(*)의 좌측에 있을 때 호출됨 <br> other 인수는 연산자 우측에 위치한 참조|
|`__floordiv__(self, other)`|정수 나눗셈(floor division). 이 메서드는 정수 나눗셈 연산자(//)의 좌측에 클래스 인스턴스가 위치할 때 호출되며, 결괏값의 소수점을 버린 정수(나눗셈의 몫)를 반환함 <br> other 인수는 연산자 우측에 위치한 참조|
|`__truediv__(self, other)`|일반 나눗셈(ordinary division). 이 메서드는 일반 나눗셈 연산자(/) 좌측에 클래스 인스턴스가 위치할 때 호출되며, 결괏값의 소수점을 버리지 않은 실수 값 그대로 반환함 <br> other 인수는 연산자 우측에 위치한 참조|
|`__divmod__(self, other)`|divmod 함수가 수행하는 나눗셈이며, 몫과 나머지 2개의 값을 지닌 튜플을 반환함 <br> other 인수는 연산자 우측에 위치한 제수(divisor) <br> 가령 divmod(17,2)를 호출하면 튜플 (8, 1)을 반환함 <br> 17을 제수 2로 나눈 몫이 8이고, 나머지가 1이기 때문|
|`__pow__(self, other)`|제곱 함수. 이 메서드는 제곱 연산자(\**)가 어떤 객체를 특정 횟수만큼 곱하려고 할 때 호출됨 <br> 가령 2**4는 숫자 2를 4번 곱하기 때문에 값이 16임 <br> other 인수는 함수에 인수로 전달되는 참조|

예를 들어 특정 클래스의 `__add__` 메서드는 한 객체가 다른 객체에 추가될 때 호출됨\
이는 객체 연산자 좌측에 위치한다고 가정할 때 이야기임\
반면 객체가 연산자 우측에 있다면 리플렉션 메서드가 호출될 것임(이경우는 `__radd__`)

다음 예제는 fractions 패키지의 Fraction 클래스를 사용한 코드

````python
import fractions

f = fractions.Fraction(1,2)
print(f + 1) # Fraction.__add__ 호출
print(2 + f) # Fraction.__radd__ 호출
````

앞서 설명했듯이 `__add__`메서드는 덧셈 연산자(+)를 인식하는 모든 클래스에 지원됨\
문자열 붙이기와 같은 연산으로 쓰일 때에도 마찬가지임

다음 가상의 Point 클래스는 산술 연산자를 위한 매직 메서드들을 어떻게 구현할 수 있을지 보여주는 예시

````python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    def __add__(self, other):
        newx = self.x + other.x
        newy = self.y + other.y
        return Point(newx, newy)
    
    def __sub__(self, other):
        dx = self.x - other.x
        dy = self.y - other.y
        
        return (dx*dx + dy*dy)**0.5
    
    def __mul__(self, n):
        newx = self.x * n
        newy = self.y * n
        
        return Point(newx, newy)
````

4개의 매직 메서드를 지원하는 이 예제 클래스는 중요한 부분을 표현하고 있음\
각 산술 연산자 메서드는 값을 반환함\
물론 이 값을 변수에 대입하거나 다른 표현식 안에서 사용할 수 있음

````python
pt1 = Point(10, 15)
pt2 = Point(0, 5)
````

표현식 pt1 + pt2는 pt1 객체의 `__add__` 메서드를 호출하게 함\
이 메서드가 호출되면 self는 pt1 객체 자체의 참조가 되며, other는 객체 pt2의 참조가 됨\
호출 결과는 신규 Point 값이 되며, 변수 x에 대입됨

이 경우 두 포인트를 더하는 것은 서로 좌표를 더하는 것이며, 결과 데이터로 신규 Point 객체가 반환됨

뺄셈 기호(-)는 이 경우 거리 연산자로 번역되며, 메서드는 두 포인트의 거리를 계산하여 하나의 부동소수점 값으로 반환함

마지막으로 곱셈 기호(*)는 좌측 피연산자(self)를 Point 객체로 가정하며 스칼라 값이 우측에 위치함\
이 메서드는 Point의 각 값을 우측 정수(n)로 곱한 신규 Point 객체를 반환함

다음 문법은 주어진 클래스의 신규 객체를 어떻게 생성하는지 다시 보여줌
```
클래스_이름(인수)
```
이 문법은 해당 클래스의 __init__ 메서드에 '인수'를 전달하여 '클래스_이름'의 신규 인스턴스를 생성함

> 클래스 인스턴스가 다른 타입과 연산을 하면서 상호 작용(interaction)한다면, 바이너리-연산 메서드는 해당 타입을 지원하지 않는 경우 반드시 NotImplemented를 반환해야만 함\
이 규칙은 연산자 우측에 위치한 피연산자에 연산 메서드를 구현할 당위성을 제공함

### 10.5 단항 산술 연산자
단항 산술 연산자 메서드는 숫자와 같이 캡슐화된 클래스에서 구현되며, 포인트, 매트릭스와 같은 다른 수학적 객체를 참조할 수 있음

|메서드 문법|설명|
|:--|:--|
|`__pos__(self)`|단항 양수(positive) 부호 <br> 이 메서드는 피연산자 하나에 덧셈 기호(+)가 있는 경우 호출됨 <br> 이 연산자는 완결성(completeness)을 보장하기 위해 제공되기 때문에 객체를 원래 상태 그대로 반환하는 것 외에는 다른 추가 작업을 수행하는 경우가 드뭄|
|`__neg__(self)`|단항 음수(negative) 부호 <br> 이 메서드는 피연산자 하나에 뺄셈 기호(-)가 있는 경우 호출됨|
|`__abs__(self)`|절댓값 <br> 클래스 객체에 abs 함수가 적용된 경우 호출됨|
|`__invert__(self)`|비트 반전(bitwise inversion) <br> 모든 비트에 위치한 1은 0으로, 0은 1로 바꾼 값을 생성함 <br> 이는 물결기호(~) 연산자 사용 시 호출됨 <br> 비트 보수(bitwise complement) 혹은 1의 보수(the one's complement)로 불리기도 함 <br> (정수의 비트 반전은 ~x로 표기하며, 결괏값은 -(x+1)임)|
|`__bool__(self)`|값을 불리언으로 변환 <br> bool()뿐만이 아니라, not이나 조건문에 응답하는 제어 구조와 같은 논리 연산자를 사용하는 경우에도 호출됨 <br> 이 메서드를 정의하지 않으면 True가 기본적으로 반환됨|
|`__round__(self, n)`|반올림 함수 <br> 이 함수는 소수점 자릿수를 제한하여 반올림하는 함수 <br> 선택적으로 추가할 수 있는 인수 n은 반올림을 몇 번째 소수점까지 할지 명시함 <br> 생략하면 가장 가까운 정수로 반올림|
|`__floor__(self)`|내림(round-down) 함수 <br> 객체의 값보다 작은 가장 큰 정수를 반환함 <br> 이 메서드는 math 패키지은 math.floor 함수에 의해 호출됨|
|`__ceil__(self)`|올림(round-upward) 함수 <br> 객체의 값보다 큰 가장 작은 정수를 반환함 <br> 이 메서드는 math 패키지의 math.ceil 함수에 의해 호출됨|
|`__trunc__(self)`|소수점 버림 함수 <br> 이 메서드는 `__floor__`와 비슷하지만 반올림을 수행하는 대신 단순히 소수점의 값을 버리면서 내림 혹은 올림을 수행함 <br> 가령 -3.5는 -3.0으로 올림이 되지만, 3.5는 3.0으로 내림이 됨 <br> 이 메서드는 math 패키지의 math.trunc 함수에 의해 호출됨|

단항 메서드를 구현하면 바이너리 메서드와 같이 신규 객체(인스턴스)를 생성하여 반환할 것이라고 기대함\
일반적으로 동일한 타입의 객체를 반환해야 함

````python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    def __add__(self, other):
        newx = self.x + other.x
        newy = self.y + other.y
        
        return Point(newx, newy)
    
    def __sub__(self, other):
        dx = self.x - other.x
        dy = self.y - other.y
        
        return (dx*dx + dy*dy)**0.5
    
    def __mul__(self, n):
        newx = self.x * n
        newy = self.y * n
        
        return Point(newx, newy)
    
    def __neg__(self):
        newx = -self.x
        newy = -self.y
        
        return Point(newx, newy)
````

예를 들어 다음 표현식은 my_point 내부 값을 변경하겠다는 의미가 아니며 대신 새로운 값을 생성하여 대입하거나 사용하겠다는 것을 의미
```
-my_point
```
이제 신규 Point 클래스 인스턴스(Point 객체)를 생성할 수 있으며 산술 음수 부호(-)를 사용할 수 있음

````python
pt1 = Point(3, 4)
pt2 = -pt1
print(pt2.x, ', ', pt2.y, sep='')
````

바라던 대로 이 결과는 3, 4로 설정한 Point 인스턴스의 산술적 음수를 출력함\
이 음수는 Point의 신규 인스턴스이며, 변수 pt2에 대입됨

매직 메서드를 직접 호출하여 테스트할 수도 있음

````python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    def __add__(self, other):
        newx = self.x + other.x
        newy = self.y + other.y
        
        return Point(newx, newy)
    
    def __sub__(self, other):
        dx = self.x - other.x
        dy = self.y - other.y
        
        return (dx*dx + dy*dy)**0.5
    
    def __mul__(self, n):
        newx = self.x * n
        newy = self.y * n
        
        return Point(newx, newy)
    
    def __neg__(self):
        newx = -self.x
        newy = -self.y
        
        return Point(newx, newy)
    
    def __trunc__(self):
        newx = self.x.__trunc__()
        newy = self.y.__trunc__()
        
        return Point(newx, newy)
````

이렇게 `__trunc__`를 정의하면 다음과 같이 메서드를 직접 테스트 할 수 있음

````python
pt1 = Point(5.5, -6.6)
pt2 = Point.__trunc__(pt1)
print(pt2.x, ', ', pt2.y, sep='')
````

### 10.6 리플렉션(역방향) 메서드
이 메서드는 객체가 표현식에서 연산자의 우측에 있을 때(혹은 두 번째 인수일 때) 호출 됨\
다음과 같이 서로 다른 클래스 2개를 더하는 표현식이 있다고 가정해 보자
```python
fido = Dog()
precious = Cat()
print(fido + precious)
```
파이썬은 표현식 fido + precious를 만나면 Dog 클래스가 `__add__` 메서드를 구현했는지 확인함\
다음과 같이 몇 가지 방법으로 동작할 것임
* 좌측 피연산자가 `__add__` 메서드를 구현했고, NotImplemented가 아닌 값을 반환함\
그러면 우측 피연산자의 어떤 메서드도 호출되지 않음

* 좌측 피연산자가(혹은 피연산자의 클래스가) `__add__` 메서드를 구현하지 않았음\
이 경우 파이썬은 우측 피연산자가 `__radd__` 메서드를 구현했는지 확인함

* 좌측 피연산자가 `__add__` 메서드를 구현했지만, 우측 피연산자와 같은 객체를 다루지 않음\
`__add__` 메서드가 우측 피연산자의 타입을 확인하고 다음과 같이 결정함\
"이 클래스의 객체에는 덧셈 연산자(+)가 지원되지 않음" $\rightarrow$ 이 경우 NotImplemented가 반환됨\
그러면 파이썬은 우측 피연산자에 `__radd__`가 구현되었는지 확인함

|메서드 문법|설명|
|:--|:--|
|`__radd__(self, other)`|덧셈. 이하 모든 메서드는 우측 피연산자가 이 메서드를 정의하고 좌측 피연산자가 해당 메서드를 정의하지 않았거나 NotImplemented를 반환하는지 확인함|
|`__rsub__(self, other)`|뺄셈|
|`__rmul__(self, other)`|곱셈|
|`__rfloordiv__(self, other)`|정수 나눗셈(floor division)|
|`__rtruediv__(self, other)`|일반 나눗셈(ordinary division)|
|`__rdivmod__(self, other)`|divmod 함수가 수행하는 나눗셈이며, 몫과 나머지 2개의 값을 지닌 튜플을 반환함|
|`__rpow__(self, other)`|제곱 함수|

대부분의 경우 역방향(reverse-order) 메서드는 정방향(forward-order)(왼쪽 피연산자) 메서드의 데칼코마니

### 10.7 교체 연산자 메서드
아래의 표는 모든 클래스에 +=, -=, *=를 포함한 산술 연산자와 대입 연산자의 조합 기능을 제공하는 매직 메서드를 나열한 것임
|메서드 문법| 설명|
|:--|:--|
|`__iadd__(self, other)`|덧셈-대입 연산자 조합 메서드 <br> 이 메서드는 클래스의 객체 += 연산자를 사용하고, 객체가 연산자의 좌측에 위치할 때 호출됨 <br> 교체 연산이 성공적으로 구현되려면 self를 반환해야 함|
|`__isub__(self, other)`|뺄셈-대입 연산자 조합 메서드 <br> 이 메서드는 클래스의 객체 -= 연산자를 사용하고, 객체가 연산자의 좌측에 위치할 때 호출됨 <br> 교체 연산이 성공적으로 구현되려면 self를 반환해야 함|
|`__imul__(self, other)`|곱셈-대입 연산자 조합 메서드 <br> 이 메서드는 클래스의 객체 *= 연산자를 사용하고, 객체가 연산자의 좌측에 위치할 때 호출됨 <br> 교체 연산이 성공적으로 구현되려면 self를 반환해야 함|
|`__idiv__(self, other)`|/= 연산 구현 <br> 이 메서드는 클래스의 객체 /= 연산자를 사용하고, 객체가 연산자의 좌측에 위치할 때 호출됨 <br> 교체 연산이 성공적으로 구현되려면 self를 반환해야 함|
|`__igrounddiv__(self, other)`|정수 나눗셈인 //= 연산 구현(가장 가까운 정수를 반환) <br> 이 메서드는 클래스의 객체 //= 연산자를 사용하고, 객체가 연산자의 좌측에 위치할 때 호출됨 <br> 교체 연산이 성공적으로 구현되려면 self를 반환해야 함|
|`__imod__(self, other)`|나머지 나눗셈 %= 연산 구현 <br> 이 메서드는 클래스의 객체 %= 연산자를 사용하고, 객체가 연산자의 좌측에 위치할 때 호출됨 <br> 교체 연산이 성공적으로 구현되려면 self를 반환해야 함|
|`__ilshift__(self, other)`|비트 좌측 시프트를 수행하는 <<= 연산 구현 <br> 이 메서드는 클래스의 객체 <<= 연산자를 사용하고, 객체가 연산자의 좌측에 위치할 때 호출됨 <br> 교체 연산이 성공적으로 구현되려면 self를 반환해야 함|
|`__irshift__(self, other)`|비트 우측 시프트를 수행하는 >>= 연산 구현 <br> 이 메서드는 클래스의 객체 >>= 연산자를 사용하고, 객체가 연산자의 좌측에 위치할 때 호출됨 <br> 교체 연산이 성공적으로 구현되려면 self를 반환해야 함|
|`__iand__(self, other)`|바이너리 AND를 수행하는 &= 연산 구현 <br> 이 메서드는 클래스의 객체 &= 연산자를 사용하고, 객체가 연산자의 좌측에 위치할 때 호출됨 <br> 교체 연산이 성공적으로 구현되려면 self를 반환해야 함|
|`__ior__(self, other)`|바이너리 OR를 수행하는 \|= 연산 구현 <br> 이 메서드는 클래스의 객체 \|= 연산자를 사용하고, 객체가 연산자의 좌측에 위치할 때 호출됨 <br> 교체 연산이 성공적으로 구현되려면 self를 반환해야 함|
|`__ixor__(self, other)`|바이너리 exclusive-OR를 수행하는 ^= 연산 구현 <br> 이 메서드는 클래스의 객체 ^= 연산자를 사용하고, 객체가 연산자의 좌측에 위치할 때 호출됨 <br> 교체 연산이 성공적으로 구현되려면 self를 반환해야 함|
|`__ipow__(self, other [, modulo])`|제곱 연산으로 불리는 **= 연산 구현 <br> 선택적 인수인 modulo가 제곱을 수행한 후 나머지 연산을 수행 <br> 이 메서드는 클래스의 객체 **= 연산자를 사용하고, 객체가 연산자의 좌측에 위치할 때 호출됨 <br> 교체 연산이 성공적으로 구현되려면 self를 반환해야 함|

여기에서 i는 'in place'를 의미함\
만약 이 메서드를 구현하면 클래스 객체의 실질적인 교체 연산이 수행되며, 메모리에 상주 중인 실제 데이터 객체의 값을 수정함

덧셈을 구현한 `__add__`와 같은 연산자는 지원하면서 이 메서드와 대응하는 i 메서드를 지원하지 않더라도, 파이썬은 여전히 대입 연산을 지원하며 이 행위는 별도의 코드 구현 없이 자동으로 제공됨\
하지만 이 연산은 메모리의 값을 교체(in place)하지 않음\
대신 신규 객체를 생성하며, 변수에 다시 대입함

예를 들어 하나의 객체와 두 번째 참조 변수가 있다고 가정해보자\
```python
a = MyClass(10)
b = a
a += 1
print(a, b)
```
여기에는 이슈가 있는데, a += 1이 교체 연산이라면 a와 b는 같은 데이터를 참조하고 있을 것이고, 함께 변경될 것임\
하지만 a += 1이 교체 연산이 아니라면 연산자는 a에 신규 데이터를 대입하고, a와 b의 연결 고리는 끊어짐\
이 경우 a와 b는 연산 후 서로 다른 데이터를 가지게 됨

문자열 클래스(str)은 변할 수 없는(immutable) 성질을 가지고 있으며, += 연산을 교체 연산자로 구현하지 않음\
대신 메모리 안에서 신규 객체를 문자열 변수에 다시 대입함

위 표의 메서드들은 실제 (메모리상의 데이터를 수정하는) 교체 연산자로 구현할 때, 다음 절차를 따라야 함\
1. 메서드가 호출된 인스턴스 내용을 수정함(self 인수로 접근하는 변수를 의미함)
2. 사용한 객체 참조를 다음과 같이 반환함
```
return self
```

예를 들어 다음과 같이 Point 클래스에 `__iadd__`와 `__imul__` 메서드를 구현할 수 있음
```python
def __iadd__(self, other):
    self.x += other.x
    self.y += other.y

    return self

def __imul__(self, other):
    self.x *= other.x
    self.y *= other.y

    return self
```

### 10.8 변환 메서드
데이터 변환 메서드는 파이썬 프로그래머들이 자주 사용하는 유용한 메서드\
예를 들어(if 문이나 while 문과 같은) 조건문에서 객체를 사용할 때, 파이썬은 암시적으로 True 혹은 False 값을 갖기 위해 bool 변환 메서드를 호출함\
이 변환 함수는 결국 객체 클래스의 `__bool__` 메서드를 호출함\
조건문에 클래스의 객체가 주어졌을 때, 어떤 값을 반환할지 결정하는 기준을 제공함\
물론 `__str__` 메서드는 문자열로 모든 것을 표현하기 때문에 변환 메서드만큼 중요함

|메서드 문법 | 설명|
|:--|:--|
|`__int__(self)`|int 변환 함수 사용 시 호출됨 <br> 이 메서드는 정수 객체를 반환해야 함|
|`__float__(self)`|float 변환 함수 사용 시 호출됨 <br> 이 메서드는 부동소수점 객체를 반환해야 함|
|`__complex__(self)`|복소수(complex)변환 함수 사용 시 호출됨 <br> 이 메서드는 복소수-숫자 객체를 반환해야 함 <br> 가령 complex(1)이 실행되면 (1+0j)를 반환함|
|`__index__(self)`|객체가 컬렉션(튜플, 문자열 혹은 리스트)의 색인이나 슬라이싱 연산의 범위로 주어진 경우 이 메서드는 실제 인덱스 숫자를 반환하며 반드시 정수여야 함|
|`__bool__(self)`|이 메서드는 적절하게 True 혹은 False를 반환해야 함 <br> 클래스 대부분은 zero 값 혹은 빈 컨테이너인 경우를 제외하고 True를 반환함|

다음 클래스 정의문은 Point 클래스에 몇 가지 변환 메서드를 정의하고 있음

````python
class Point:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y
    
    def __int__(self):
        return int(self.x) + int(self.y)
    
    def __float__(self):
        return float(self.x) + float(self.y)
````

````python
p = Point(1, 2.5)
int(p)
````

````python
float(p)
````

### 10.9 컬렉션 클래스 메서드
파이썬에서는 사용자가 원하는 방식대로 저장되는 자체 컬렉션 클래스를 생성할 수 있음\
예를 들어 해시 테이블 대신 바이너리 트리 기반으로 구현한 딕셔너리 클래스를 생성할 수 있음\
또한 기존 파이썬 컬렉션 클래스 기반에 부가 기능을 추가한 자체 컬렉션 클래스를 생성할 수 있음
아래의 표는 자체 컬렉션 클래스를 생성할 때 제공해야 하는 매직 메서드 리스트임

|메서드 문법|설명|
|:--|:--|
|`__len__(self)`|컬렉션 내에 항목의 개수를 정수로 반환함|
|`__getitem__(self, key)`|컬렉션에서 주어진 색인의 항목을 반환함 <br> 이 매직 메서드는 객체이름[색인] 형태로 색인을 사용할 때 호출됨 <br> 색인이 정수이거나 데이터 딕셔너리와 같이 정수가 아닐 수도 있지만, 양쪽 모두 컬렉션으로부터 선택된 항목을 반환함|
|`__setitem__(self, key, value)`|이 매직 메서드의 목적은 색인 위치의 값을 특정 값으로 설정하는 것을 제외하고는 `__getitem__`과 유사함 <br> 여기에서 색인은 가령 색인이 정수인 경우 정수 숫자임 <br> 색인에 의해 선택된 항목은 특정 값으로 교체됨 <br> 기존 객체 안에 데이터를 바로 수정함|
|`__delitem__(self, key)`|이 메서드는 색인으로 선택한 특정 항목을 삭제함|
|`__iter__(self)`|컬렉션 객체의 이터레이터(iterator)를 반환함 <br> 이 이터레이터는 컬렉션의 next 메서드를 구현한 객체 <br> 그러므로 `__iter__` 메서드를 구현하는 가장 쉬운 방법은 `__next__` 메서드를 직접 구현한 클래스의 self를 반환하는 것|
|`__next__(self)`|이 메서드는 클래스 스스로 혹은 메인 클래스와 함께 동작하는 조력(helper) 클래스에 의해 구현될 수 있음 <br> 두 경우 모두 순회 시 다음 항목을 반환하거나 StopIteration 예외를 발생해야 함|
|`__reversed__(self)`|이 메서드는 컬렉션의 항목을 역방향으로 순회하는 이터레이터를 반환해야 함|
|`__contains__(self, item)`|이 메서드는 컬렉션 안에 특정 항목의 존재 유무에 따라 True 혹은 False를 반환해야 함|
|`__missing__(self, key)`|이 메서드는 컬렉션에 존재하지 않는 항목에 접근할 때 호출됨 <br> 경우에 따라서 None을 반환하거나 예외를 발생시킬 수도 있음|

다음 코드는 간단한 컬렉션 클래스 예제인데, 특정 용도로 만든 스택 클래스로, 이미 존재하는 함수에 기능을 추가했음\
가령 peek 함수는 스택에서 항목을 끄집어 내지 않고 항목 중 가장 '높은'(혹은 마지막) 항목을 반환함

````python
class Stack:
    def __init__(self):
        self.mylist=[]
        
    def append(self, v):
        self.mylist.append(v)
    
    def push(self, v):
        self.mylist.append(v)
    
    def pop(self):
        return self.mylist.pop()
    
    def peek(self):
        return self.mylist[-1]
    
    def __len__(self):
        return len(self.mylist)
    
    def __contains__(self, v):
        return self.mylist.__contains__(v) # 포함(containment)은 여기에서 사용됨
    
    def __getitem__(self, k):
        return self.mylist[k]
````

이 클래스 정의문으로 스택 객체를 생성하여 컬렉션처럼 다룰 수 있음

````python
st = Stack()
st.push(10)
st.push(20)
st.push(30)
st.push(40)
print(f'Size of stack is {len(st)}')
print(f'First element is {st[0]}')
print(f'The top of the stack is {st.peek()}')
print(st.pop())
print(st.pop())
print(st.pop())
print(f'Size of stack is {len(st)}')
````

객체 지향 프로그래밍에 익숙하다면 이런 상황에서 상속을 사용하면 더 쉽게 같은 결과를 얻을 수 있다는 것을 알 수 있음

````python
class Stack(list):
    def push(self, v):
        self.append(v)
    
    def peek(self):
        return self[-1]
````

몇 줄의 코드로 Stack 클래스는 앞서 살펴본 것보다 더 정교한 클래스 정의문의 모든 연산자를 다룰 수 있게 되었음\
상속과 같은 방법은 리스트나 딕셔너리와 같이 이미 존재하는 클래스를 기반으로 자체 컬렉션 클래스를 만들 때 사용할 수 있음

### 10.10 `__iter__`와 `__next__`구현하기
next 함수와 함께 `__iter__`와 `__next__` 메서드는 정교하게 다룰 필요가 있음\
이 메서드들은 for 루프와 같은 특별한 상황에 사용할 수 있는 클래스 객체의 제너레이터(generator)(혹은 이터레이터)를 생성함

* 이터러블(iterable)은 여러 항목을 한 번에 한 항목씩 접근하여 첫 항목부터 끝 항목까지 관통(step through)할 수 있는 객체를 의미\
객체가 이터러블이 되기 위해 `__iter__` 메서드는 반드시 객체를 반환해야 함

* 이터레이터(iterator)는 `__iter__`가 반드시 반환해야 하는 객체임\
이터레이터는 컬렉션 객체를 관통하는 데 사용됨

* 이터레이터는 `__next__` 메서드가 반환해야 하는 객체이며, 동일한 클래스이거나 이 목적만을 위해 작성된 별도 클래스일 수도 있음\
`__next__` 메서드는 아무 작업을 하지 않더라도 구현되어야 함

이 메서드의 구현은 파이썬의 for문에서 클래스 인스턴스를 사용하려면 반드시 필요함\
예를 들어 4차원의 Point 객체가 있다고 해 보자\
이터레이터가 4 좌표 값을 한 번에 하나씩 가져올 수 있다면 다음과 같이 사용할 수 있음
```python
my_point=Point()
for i in my_point:
    print(i)
```
이 예제는 좌표 4개를 한 번에 하나씩 출력함\
컬렉션은 그 자체가 이터레이터일 수 있음\
그런 경우라면 `__iter__` 메서드는 self를 반환하며, 자체적으로 `__next__` 메서드를 구현했을 것임\
컨테이너의 복잡성과 원하는 유연성 수준에 따라서 이 메서드들을 여러 가지 방법으로 구현할 수 있음

* 대상 내에 저장된 컬렉션 객체의 `__iter__`를 호출함\
가장 쉬운 방법이며, 다른 누군가는 순회 작업을 제어해야 함

* 컬렉션 클래스 자체적으로 `__iter__`와 `__next__`를 모두 구현함\
이 경우에 `__iter__` 메서드는 이터레이션 세팅을 초기화하며, self를 반환함\
반면 이 방법으로는 한 번에 한 루프 이상 순회할 수가 없음

* `__iter__` 메서드가 컬렉션 클래스를 관통하는 이터레이션을 지원하는 목적으로 만들어진 자체 이터레이터 객체를 반환함\
이 방법이 가장 강력하며, 시도해 볼 만함

설명한 방법 중 두 번째 방법으로 구현한 코드며, 일반적으로 사용하는 비교적 간단한 예제

````python
class Stack:
    def __init__(self):
        self.mylist=[]
        
    def append(self, v):
        self.mylist.append(v)
    
    def push(self, v):
        self.mylist.append(v)
    
    def pop(self):
        return self.mylist.pop()
    
    def peek(self):
        return self.mylist[-1]
    
    def __len__(self):
        return len(self.mylist)
    
    def __contains__(self, v):
        return self.mylist.__contains__(v) # 포함(containment)은 여기에서 사용됨
    
    def __getitem__(self, k):
        return self.mylist[k]
    
    def __iter__(self):
        self.current = 0
        return self
    
    def __next__(self):
        if self.current < len(self):
            self.current += 1
            return self.mylist[self.current - 1]
        else:
            raise StopIteration
````

이 코드의 중요한 코딩 기법은 self.current에서 사용한 current 변수를 참조하는 방법임\
이는 변수가 클래스나 글로벌 변수가 아닌 인스턴스 변수가 되게 만듬

모든 항목이 차례로 나열되거나 순회할 때 계속해서 self.current를 증가시키면서 `__new__` 메서드가 StopIteration 예외를 일으킬 때까지 컬렉션을 관통함\
StopIteration은 for 루프를 멈추게 함

## 11. 다중 인수 타입 지원
하나 이상의 인수를 가지는 함수나 메서드를 어떻게 작성할 수 있을까?\
예를 들어 Point 객체가 스칼라 숫자나 다른 Point 객체와 곱하는 Point 클래스가 있다고 가정해 보자\
파이썬은 오버로딩(overloading)을 지원하지 않지만 런타임 시 인수의 타입을 테스트할 수 있으며, 타입에 따라서 다르게 동작할 수 있음(오버로딩은 인수를 다르게 하여 여러 번 정의하는 것을 의미함)

객체 타입을 테스트하는 방법 중 하나는 type 함수를 호출하는 것
```python
type(객체)
```
가령 데이터 객체나 변수가 정수 타입을 가지고 있는지 직접 확인할 수 있음

````python
n = 5
if type(n) == int:
    print('n is integer.')
````

다른 방법으로는 isinstance 함수를 사용하는 것임
```python
isinstance(객체, 클래스_이름)
isinstance(객체, 클래스들로 구성된 튜플)
```
첫 번째 버전은 '객체'의 클래스를 확인한 후 이 클래스가 '클래스_이름(두 번째 인수)'과 동일하거나 '클래스_이름'에서 파생된 클래스면 True를 반환함

두 번째 버전은 위 문법과 동일하지만 두 번째 인수에 여러 클래스로 구성된 튜플(변경할 수 없는 리스트)의 클래스 중 하나라도 동일 조건에 만족하면 True를 반환함

다음 코드는 첫 번째 문법을 사용한 예시

````python
n = 5
if isinstance(n, int):
    print('n is an integer or derived from it.')
````

두 번째 문법의 예제도 살펴보자.(이 방식은 n이 정수나 부동소수점 숫자를 포함하는지 테스트함)

````python
if isinstance(n, (int, float)):
    print('n is numeric.')
````

type과는 다르게 isinstance를 사용했기 때문에 n이 int 혹은 float 타입을 반드시 가질 필요가 없음\
int 혹은 float로부터 파생된 타입이어도 충분하며 언제든지 하위 클래스를 만들 수 있음

예를 들어 Point 객체가 다른 Point 객체와 숫자를 곱할 수 있게 만든다고 가정해 보자\
다음과 같이 `__mul__`메서드를 구현하면 됨
```python
def __mul__(self, other):
    if type(other)==Point:
        newx = self.x * other.x
        newy = self.y * other.y
        return Point(newx, newy)
    elif type(other) == int or type(other) == float:
        newx = self.x * other
        newy = self.x * other
        return Point(newx, newy)
    else:
        return NotImplemented
```
여기에서 other 클래스를 포함한 연산 처리를 어떻게 해야 할지 모를 때 예외를 일으키는 대신 NotImplemented를 반환하는 것에 주목\
NotImplemented를 반환하면, 파이썬은 우측 피연산자의 클래스가 이 상황을 제어할 `__rmul__` 메서드가 있는지 확인할 것임\
이 메서드는 타입 확인을 위해 type 함수 대신 isinstance 함수를 사용하여 정의할 수도 있음
```python
def __mul__(self, other):
    if isinstance(other, Point):
        newx = self.x * other.x
        newy = self.y * other.y
        return Point(newx, newy)
    elif isinstance(other, (int, float)):
        newx = self.x * other
        newy = self.x * other
        return Point(newx, newy)
    else:
        return NotImplemented
```
두 경우 모두, 숫자에 의한 곱셈은 비대칭으로 수행됨(`__rmul__`메서드를 구현하지 않았기 때문에 비대칭으로 표현)\
다음과 같이 작성한 코드를 실행하면 확인할 수 있음
```python
pt2 = 5.5 * pt1
```
정수나 부동소수점 클래스는 우리가 만든 것이 아니기 때문에 당연히 매직 메서드가 추가되지 않았으며, 이것이 문제가 됨\
이 문제를 해결하려면 Point 클래스에 `__rmul__` 메서드를 반드시 작성해야 함
```python
def __rmul__(self, other):
    if isinstance(other, (int, float)):
        newx = self.x * other
        newy = self.x * other
        return Point(newx, newy)
    else:
        return NotImplemented
```

## 12. 동적 속성 설정 및 조회
파이썬 객체는 수많은 속성(attribute)을 가질 수 있음\
속성의 범주에는 인스턴스 변수, 메서드와 인수들(property)을 포함함\
일반적으로 이런 속성들은 하드 코딩되어 있어서 프로그래밍 코드 안에서 이름이 정해져 있음\
하지만 때때로 동적으로 속성을 설정(setting)하는 것이 유용할 때가 존재함\
프로그램 실행 시점에 특정 조건에 따라 속성 이름이 결정되는 것임\
예를 들어 사용자가 속성 이름을 제안하거나, 혹은 데이터베이스나 다른 애플리케이션에서 속성 이름을 설정하는 경우가 존재함

setattr 함수는 다음 문법을 따름
```python
setattr(객체_이름, 문자열_이름, 값)
```
인수를 살펴보면 '객체_이름'이 수정 대상 객체의 참조고, '문자열_이름'은 속성 이름을 명시한 문자열이며, '값'은 데이터 객체이거나 값을 가지고 있는 표현식임

getattr 함수는 setattr과 상호 보완적인 문법을 가지고 있음
```python
getattr(객체, 문자열_이름 [,기본값])
```
다음의 간단한 예제를 IDLE 환경에서 실행해 보자\
속성 breed는 동적으로 추가되며, 'Great Dane'으로 설정됨

````python
class Dog:
    pass
d = Dog()
setattr(d, 'breed', 'Great Dane')
getattr(d, 'breed')
````

하지만 실제로 getattr을 사용할 때는 다음과 같이 대부분 문자열을 담고 있는 변수를 인수로 전달함

````python
field = 'breed'
getattr(d, field)
````

파이썬은 객체 지향 프로그래밍에서 추구하는 유연하고 강력한 기능을 제공함\
클래스의 기본 개념은 근본적인 사용자-정의 타입이지만, 사용자가 원하는 만큼의 메서드를 정의할 수 있음\
메서드는 클래스 안에 정의된 함수이며, 일반적으로 클래스의 인스턴스를 통해 호출됨
```python
my_dog = Dog('Bowser')
my_dog.speak()
```
파이썬 메서드는 일반적으로 self로 불리는 변수를 첫 번째 인수로 배치함\
인수 self는 메서드 호출 시 명시적으로 인수에 넣지 않음\
하지만 각 인스턴스를 통해 호출될 메서드 정의문에는 항상 나타나야 함\
self는 해당 객체 스스로의 참조

파이썬은 실행 시점까지(문장이 실행되기 전까지) 변수와 함수의 이름이 확정되지 않기 때문에 극도로 다형적임\
여러 클래스에서 동일한 이름의 속성을 정의할 수 있지만, 생성된 특정 객체를 참조하기 위해 작성된 코드는 항상 의도대로 동작할 수 있음

파이썬의 가장 독특한 기능은 모든 클래스가 매직 메서드를 이용할 수 있다는 것임\
메서드 이름은 파이썬에서 특별한 의미를 가지고 있으며, 특정 상황에 따라 자동으로 호출됨\
가령 `__init__` 메서드는 클래스의 인스턴스가 초기화될 때 호출됨\
매직 메서드 이름은 항상 언더스코어 2개가 앞뒤로 붙음\
따라서 이런 방식의 이름을 사용하지 않으면 매직 메서드와 이름이 겹치는 것을 피할 수 있음
