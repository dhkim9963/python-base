## 3.1 데이터 프레임
데이터 프레임(data frame)은 다소 생소한 개념으로 느껴지지만, 일상생활에서 흔히 접하는 표와 동일한 클래스\
엑셀이나 구글 스프레드시트 등에서 사용하는 2차원 표를 판다스는 데이터 프레임으로 다룸

### 3.1.1 데이터 프레임이란?
우리는 데이터를 표 형식으로 처리하며 판다스에서 표 데이터를 다루는 클래스가 데이터 프레임이며 크게 3가지 부분으로 이루어짐
* 인덱스(index) : 행 이름의 집합
* 컬럼즈(columns) : 열 이름의 집합
* 밸류즈(values) : 데이터에 해당하는 점수의 집합

````python
import pandas as pd # 판다스 라이브러리 불러오기
df = pd.DataFrame([[63, 93, 97], [89, 83, 71], [83, 76, 92], [94, 88, 73]],
                  index=['김판다', '강승주', '송중기', '권보아'],
                  columns=['국어', '영어', '수학'])
df
````

파이썬에서는 일반적으로 사용하는 리스트나 넘파이의 어레이 등의 클래스는 키를 기반으로 하는 인덱싱을 지원하지 않음\
김판다의 영어 점수를 가져오려면 0번 행(첫 번째 행)과 1번 열(두 번째 열)의 값을 인덱싱해야 함\
로케이션은 맨 처음 위치에 0을 부여하기에 두 번째 열의 로케이션은 1임\
로케이션을 활용한 인덱싱도 편리하나, 위치를 기반으로 하므로 데이터의 의미를 파악하기 어렵다는 단점이 존재함

데이터 프레임은 로케이션 인덱싱을 지원하지만, 행과 열의 이름을 키로 사용하는 키 인덱싱도 함께 지원함\
로케이션을 활용한 df.iloc[0, 1]로도 인덱싱이 가능하지만, 키를 활용한 df.loc['김판다', '영어']로 인덱싱도 가능함\
데이터 프레임이 딕셔너리의 장점인 키 기반 인덱싱과 리스트와 어레이의 장점인 로케이션 기반 인덱싱을 모두 지원한다는 의미임\
대부분의 클래스가 키나 로케이션 둘 중 하나만 사용하는 것과 달리 판다스의 데이터 프레임과 시리즈는 두 가지를 모두 사용함

인덱스와 컬럼즈가 데이터에 접근하는 색인이라면, 밸류즈는 실제 데이터를 담는 부분임\
데이터 프레임은 2차원 구조이므로 데이터가 모인 2차원 배열을 밸류즈라고 함(시리즈는 1차원 구조 이므로 밸류즈가 1차원 배열이 됨)\
데이터 프레임의 밸류즈 부분은 넘파이의 어레이 클래스인데, 판다스 라이브러리가 넘파이 라이브러리를 기반으로 만들어졌기 때문

밸류즈가 넘파이의 어레이로 이루어지기에 판다스의 데이터 프레임은 어레이의 많은 부분을 계승해 유사한 특징이 있음\
대표적으로 비교 연산을 통해 데이터 프레임이 손쉽게 불 자료형 데이터 프레임으로 변환하는 것은 어레이의 특징인 벡터화 연산을 계승한 것

````python
df>90 # 데이터를 담고 있는 밸류즈 부분이 비교 연산으로 손쉽게 True, False로 변환됨
````

### 3.1.3 데이터 프레임 생성하기
데이터 프레임은 DataFrame 함수로 생성함\
데이터 프레임이 인덱스, 컬럼즈, 밸류즈의 세 부분으로 구성되어, 데이터 프레임을 코드로 생서앟ㄹ 때도 세 가지를 모두 지정해야 함\
생성할 때 인덱스와 컬럼즈를 별도로 지정하지 않으면 0부터 순차적으로 정수를 부여하며, 이 인덱스 클래스를 RangeIndex라고 함

````python
import pandas as pd
data1 = [[63, 93, 97], [89, 83, 71], [83, 76, 92], [94, 88, 73]]
df = pd.DataFrame(data1)
df
````

````python
df = pd.DataFrame(data1, index=['김판다', '강승주', '송중기', '권보아'])
df
````

DataFrame 함수의 index 매개변수를 활용하면 원하는 인덱스가 설정됨\
원하는 컬럼즈를 설정하려면 매개변수 columns에 설정하고 싶은 배열을 입력함

````python
df = pd.DataFrame(data1, index=['김판다', '강승주', '송중기', '권보아'],
                  columns=['국어', '영어', '수학'])
df
````

데이터 프레임은 딕셔너리를 사용하여 생성할 수 있음\
이때는 밸류즈, 인덱스, 컬럼즈가 모두 포함된 2차원 딕셔너리를 생성하여 DataFrame 함수에 입력함\
컬럼즈가 상위 딕셔너리의 키(key)가 되고, 하위 딕셔너리에 인덱스와 데이터가 키와 밸류로 입력되는 2차원 구조

````python
data2 = {'국어': {'김판다': 63, '강승주': 89, '송중기': 83, '권보아': 94},
         '영어': {'김판다': 93, '강승주': 83, '송중기': 76, '권보아': 88},
         '수학': {'김판다': 97, '강승주': 71, '송중기': 92, '권보아': 73}}
df = pd.DataFrame(data2)
df
````

### 3.1.4 매개변수와 인수, 기본값
$f(x, y) = 2x + y $ 일 때 $f(2, 1)$의 값은?\
위 문제는 어렵지 않게 풀 것인데, x의 자리에 2를 넣고 y의 자리에 1을 넣어서 정답 5를 도출할 것임\
함수 f에서 x와 y는 함수의 입력값을 나타내는 변수로, 함수를 정의할 때 값이 입력될 자리를 만들어둔 것이며 이 x,y를 매개변수라 함\
$f(2, 1)$에서 2와 1은 함수의 입력값으로 매개변수 x와 y에 전달되는 값이며 매개변수에 전달되는 입력값이 인수임\
위 예시에서 매개변수 x의 인수는 2, y의 인수는 1임

코딩에서 매개변수와 인수를 동시에 표기할 때가 많은 이유는 인수만 입력할 때는 매개변수를 입력하지 않아도 되어 간편하지만, 순서를 엄격히 따라 모든 인수를 입력해야 하는 단점이 존재하기 때문\
매개변수를 인수와 함께 입력하면 순서에 구애받지 않고, 일부의 매개변수에만 인수를 입력해서 함수를 사용할 수 있음\
내가 인수를 입력하지 않은 매개변수에도 인수는 전달되어야 하기에 함수를 생성할 때 인수를 입력하지 않아도 매개변수에 전달될 값을 정의함\
별도의 인수를 입력하지 않아도 전달되는 인수를 기본값이라고 함\
위의 데이터프레임 예시에서 매개변수 index와 columns에 인수를 입력하지 않아 기본값으로 RangeIndex가 설정된 이유임\
매개변수의 순서에 맞게 인수를 입력하면 매개변수를 생략할 수 있는데, 너무 많이 생략하면 코드 가독성이 떨어지는 경우가 발생하므로 첫 번째 매개변수를 처음 입력할 때 등 몇몇 경우에만 매개변수를 생략함

## 3.4 데이터 프레임 저장하기
코드로 데이터 프레임을 생성하거나 파일에서 불러온 데이터 프레임을 기반으로 새로운 데이터 프레임을 생성한 후, 그 겨로가물을 다른 형태로 변환해야 할 때가 있는데, 데이터 프레임을 딕셔너리로 변환하거나 파일로 저장하는 방법을 살펴 봄

### 3.4.1 딕셔너리로 변환하기(to_dict)
데이터 프레임은 파이썬의 판다스에서만 사용 가능한 데이터 구조이기에 다른 프로그래밍 언어로 결과 데이터를 활용하고자 할 때나 파이썬 내에서 판다스가 아닌 다른 라이브러리로 데이터를 다룰 때도 주로 딕셔너리로 변환함

데이터 프레임은 to_dict 함수로 딕셔너리로 변환함

````python
df.to_dict()
````

### 3.4.2 파일로 저장하기(to_excel, to_csv)
데이터 프레임인 변수 ㅇf를 파일로 저장할 것인데 먼저 엑셀 파일로 저장할 때 to_excel 메서드를 사용함

````python
df.to_excel('1.xlsx')
````

위처럼 저장하면 인덱스도 함께 저장되는데 RangeIndex는 인덱스가 특별한 데이터를 갖지 않으므로 인덱스는 제외하고 파일로 저장해야함\
to_excel 메서드의 매개변수 index에 False를 입력하면 인덱스는 제외하고 파일로 저장됨

````python
df.to_excel('1.xlsx', index=False)
````

csv 파일로 저장하려면 to_csv 함수에 파일명을 인수로 입력함\
기본값으로 구분자는 쉼표로 지정되는데, 2차원 표 데이터를 쉼표로 열을 구분하고 줄 바꿈으로 행을 구분해 텍스트로 저장함\
인코딩 방식은 기본값으로 UTF-8로 지정됨

````python
df.to_csv('2.csv', index=False)
````

### 3.4.3 함수와 메서드, 속성
데이터 프레임을 저장하고자 사용한 to_dict, to_excel, to_csv 함수는 모두 메서드인데 함수와 메서드의 차이에 대해 살펴보자

메서드는 함수(function)의 일종이라서 매개변수에 인수를 전달받아 값을 반환함\
그러나 함수는 독립적으로 호출할 수 있는 반면, 메서드는 클래스 내부에 정의되어 객체에 종속되는 함수이기에 특정 객체에 종속되어 호출됨\
함수는 객체와 관계없이 호출할 수 있지만, 메서드는 객체를 먼저 불러온 후 호출해야 하므로 메서드는 객체 뒤에 붙여 사용함

1. pd.read_excel('파일경로명') : 함수, 판다스 함수는 앞에 pd.를 붙여 사용하며, 인수를 입력받아 결과를 반환함
2. df.to_dict() : 메서드, 객체에 종속되어 호출되므로 객체(df)를 불러온 뒤 메서드를 객체 뒤에 붙여 사용함
3. df.index : 속성, 속성은 클래스 내부에서 정의된 변수로 객체 뒤에 붙여 사용하고 결과를 반환하지만 함수가 아니라 인수를 받지 않음

속성과 메서드도 혼동하기 쉬운데, 속성은 메서드처럼 객체 뒤엥 붙여 사용하지만 클래스 내부에 정의된 변수를 반환하므로 인수가 입력되지 않음\
대부분의 함수를 메서드 형태로 사용할 것인데, 판다스에서 함수로 표기하더라도 대부분은 메서드를 지칭함\
함수보다 메서드를 주로 사용하는 이유는 판다스는 하나의 객체에 여러 가지 함수를 연속해서 사용하는 작업이 많기 때문임\
그래서 함수를 여러 번 사용하기보다 메서드를 연속해서 사용하여(chain method) 코드를 작성하면, 가독성이 높아지고 유지보수가 쉬움\
```py
연속함수 pd.funcC(pd.funcB(pd.funcA(df)))
연속 메서드 df.funcA().funcB().funcC() # funcA, funcB, funcC의 순차적 사용을 한눈에 파악하기 쉬움
```

대괄호 인덱서는 인덱싱과 슬라이싱 모두 지원하지만 열의 인덱싱과 행의 슬라이싱만 지원함

````python
df['B':'D']
````

대괄호 인덱서는 행만 슬라이싱 가능하다는 제한으로 인해 슬라이싱할 때 잘 사용되지 않음\
불리언 인덱싱에서 대괄호 인덱서를 자주 사용하므로 대괄호 인덱서의 슬라이싱에 대해 개념은 파악하는 것이 좋음

### 4.1.5 키 인덱싱과 슬라이싱(loc 인덱서)
만약 A 학새으이 국어 성적이 궁금하다면?\
이런 작업은 대괄호 인덱싱으로는 두 번의 대괄호 인덱싱으로 추출해야 함\
대괄호 인덱싱은 2차원 구조인 데이터 프레임의 인덱싱을 완전히 만족시킬 수 없기에 간단한 인덱싱에 사용됨

대괄호 인덱싱과 달리 loc 인덱서와 iloc 인덱서는 모두 행과 열을 동시에 인덱싱할 수 있음\
loc 인덱서는 (key) location의 약자로서, 키를 기반으로 인덱싱을 수행하므로 열 이름을 입력함\
iloc 인덱서는 integer location의 약자로, 위치를 정수로 부여한 로케이션을 기반으로 인덱싱을 수행하므로 로케이션을 입력함

````python
df2 = df.copy()
df2.loc['A', '국어']
````

loc 인덱서는 행과 열을 동시에 인덱싱할 수 있기에 셀의 데이터를 추출할 수 있음\
loc 인덱서 역시 복수 인덱싱이 가능한데, 복수일 때는 이름들을 리스트로 입력해야 함

````python
df2.loc[['A', 'C', 'D'], ['국어', '수학']]
````

loc 인덱서도 슬라이싱이 가능함

````python
df2.loc['B':'D', '영어':'과학']
````

슬라이싱할 때 시작점이 데이터 프레임의 처음이거나, 끝 지점이 데이터 프레임의 마지막이라면 생략할 수 있음

````python
df2.loc['A':'D', '수학':'불어']
````

````python
df2.loc[:'D', '수학':]
````

데이터 프레임의 처음부터 끝까지 슬라이싱할 때, 시작점과 끝 지점을 모두 생략할 수 있는데 행의 슬라이싱에서는 콜론만 입력하면 됨

````python
df2.loc['A':'E', '수학':'과학']
````

````python
df2.loc[:,'수학':'과학']
````

모든 열을 가져올 때는 열의 슬라이싱 전체를 생략할 수 있으며 쉼표를 사용하지 않고 행의 슬라이싱만 입력함

````python
df2.loc[:'D',:]
````

````python
df2.loc[:'D']
````

하지만 행을 전부 가져올 때는 행의 슬라이싱은 생략될 수 없음

````python
df2.loc[:, '수학':'과학']
````

````python
df2.loc['수학':'과학']
````

행은 인덱싱을 수행하지만 열은 슬라이싱을 수행할 수 있고, 행은 슬라이싱을 수행하지만 열은 인덱싱을 수행할 수 있음

````python
df2.loc[:'D',['과학','수학']]
````

### 4.1.6 loc 인덱서로 행과 열 생성하기
열 생성은 대괄호 인덱서를 사용하는 것이 간편하므로 일반적으로 열의 생성은 대괄호 인덱서를 사용하는 것이 바람직함\
일부에만 값을 부여하는 열을 생성할 때는 loc 인덱서를 사용하는 것이 효과적임

````python
df2.loc[['B','D'],'실험'] = 80
df2
````

주로 빅데이터처럼 대규모 데이터를 다루므로 행 이름을 일일이 입력하여 특정 행에만 값을 부여하는 열을 생성하지는 않지만 조건에 따라 값을 다르게 부여하는 열을 생성할 때 이 방식을 사용함\
열 생성 외에 셀의 데이터도 수정할 수 있음

````python
df2.loc['A','과학']= 70
df2
````

loc 인덱서로는 행도 생성할 수 있는데, 존재하지 않는 행을 인덱싱해 값을 부여하면 새로운 행이 생성되며 데이터 프레임의 마지막에 위치함

````python
df2.loc['F'] = 60
df2
````

### 4.1.7 로케이션 인덱싱과 슬라이싱(iloc 인덱서)
데이터 프레임은 로케이션으로 인덱싱과 슬라이싱이 가능하며 이때는 iloc 인덱서를 활용함\
로케이션은 가장 처음의 원소에 0을 부여하고, 그 뒤로 1씩 순차적으로 증가하는 정수를 부여함

````python
df3 = df.copy()
df3.iloc[0,0]
````

iloc 인덱서 역시 복수의 행과 열을 인덱싱하는 것도 가능한데, 복수의 행과 열을 인덱싱할 때는 로케이션을 리스트로 묶어 입력함

````python
df3.iloc[[0, 2, 3], [0, 2]]
````

iloc 인덱서도 슬라이싱을 수행하며 시작점과 끝 지점의 로케이션을 콜론으로 구분해서 입력함

````python
df3.iloc[1:3, 1:4]
````

loc 인덱서는 양쪽 경계를 모두 포함하지만, iloc 인덱서는 시작점은 포함하고 끝지점은 포함하지 않음\
이는 리스트의 슬라이싱과 동일한 방식이며, 파이썬의 로케이션 슬라이싱이 가지는 공통적인 특징임\
iloc 인덱서도 슬라이싱할 때 시작점이 데이터 프레임의 처음이거나, 끝 지점이 데이터 프레임의 마지막이면 생략할 수 있음

````python
df3.iloc[:4, 2:]
````

행을 모두 가져올 때는 시작점과 끝 지점이 모두 생략되어 콜론만으로 표현할 수 있음\
열을 모두 가져올 때는 열의 슬라이싱을 전체적으로 생략할 수 있음

````python
df3.iloc[0]
````

````python
df3.iloc[-1]
````

또한 인덱싱과 슬라이싱은 행과 열에 각각 적용되며, 인덱싱이 순서에 구애받지 않는다는 점은 iloc 인덱서도 갖는 특징임

````python
df3.iloc[1:4, [-2, 3, -2]]
````
