## 3.1 데이터 프레임
데이터 프레임(data frame)은 다소 생소한 개념으로 느껴지지만, 일상생활에서 흔히 접하는 표와 동일한 클래스\
엑셀이나 구글 스프레드시트 등에서 사용하는 2차원 표를 판다스는 데이터 프레임으로 다룸

### 3.1.1 데이터 프레임이란?
우리는 데이터를 표 형식으로 처리하며 판다스에서 표 데이터를 다루는 클래스가 데이터 프레임이며 크게 3가지 부분으로 이루어짐
* 인덱스(index) : 행 이름의 집합
* 컬럼즈(columns) : 열 이름의 집합
* 밸류즈(values) : 데이터에 해당하는 점수의 집합

````python
import pandas as pd # 판다스 라이브러리 불러오기
df = pd.DataFrame([[63, 93, 97], [89, 83, 71], [83, 76, 92], [94, 88, 73]],
                  index=['김판다', '강승주', '송중기', '권보아'],
                  columns=['국어', '영어', '수학'])
df
````

파이썬에서는 일반적으로 사용하는 리스트나 넘파이의 어레이 등의 클래스는 키를 기반으로 하는 인덱싱을 지원하지 않음\
김판다의 영어 점수를 가져오려면 0번 행(첫 번째 행)과 1번 열(두 번째 열)의 값을 인덱싱해야 함\
로케이션은 맨 처음 위치에 0을 부여하기에 두 번째 열의 로케이션은 1임\
로케이션을 활용한 인덱싱도 편리하나, 위치를 기반으로 하므로 데이터의 의미를 파악하기 어렵다는 단점이 존재함

데이터 프레임은 로케이션 인덱싱을 지원하지만, 행과 열의 이름을 키로 사용하는 키 인덱싱도 함께 지원함\
로케이션을 활용한 df.iloc[0, 1]로도 인덱싱이 가능하지만, 키를 활용한 df.loc['김판다', '영어']로 인덱싱도 가능함\
데이터 프레임이 딕셔너리의 장점인 키 기반 인덱싱과 리스트와 어레이의 장점인 로케이션 기반 인덱싱을 모두 지원한다는 의미임\
대부분의 클래스가 키나 로케이션 둘 중 하나만 사용하는 것과 달리 판다스의 데이터 프레임과 시리즈는 두 가지를 모두 사용함

인덱스와 컬럼즈가 데이터에 접근하는 색인이라면, 밸류즈는 실제 데이터를 담는 부분임\
데이터 프레임은 2차원 구조이므로 데이터가 모인 2차원 배열을 밸류즈라고 함(시리즈는 1차원 구조 이므로 밸류즈가 1차원 배열이 됨)\
데이터 프레임의 밸류즈 부분은 넘파이의 어레이 클래스인데, 판다스 라이브러리가 넘파이 라이브러리를 기반으로 만들어졌기 때문

밸류즈가 넘파이의 어레이로 이루어지기에 판다스의 데이터 프레임은 어레이의 많은 부분을 계승해 유사한 특징이 있음\
대표적으로 비교 연산을 통해 데이터 프레임이 손쉽게 불 자료형 데이터 프레임으로 변환하는 것은 어레이의 특징인 벡터화 연산을 계승한 것

````python
df>90 # 데이터를 담고 있는 밸류즈 부분이 비교 연산으로 손쉽게 True, False로 변환됨
````

### 3.1.3 데이터 프레임 생성하기
데이터 프레임은 DataFrame 함수로 생성함\
데이터 프레임이 인덱스, 컬럼즈, 밸류즈의 세 부분으로 구성되어, 데이터 프레임을 코드로 생서앟ㄹ 때도 세 가지를 모두 지정해야 함\
생성할 때 인덱스와 컬럼즈를 별도로 지정하지 않으면 0부터 순차적으로 정수를 부여하며, 이 인덱스 클래스를 RangeIndex라고 함

````python
import pandas as pd
data1 = [[63, 93, 97], [89, 83, 71], [83, 76, 92], [94, 88, 73]]
df = pd.DataFrame(data1)
df
````

````python
df = pd.DataFrame(data1, index=['김판다', '강승주', '송중기', '권보아'])
df
````

DataFrame 함수의 index 매개변수를 활용하면 원하는 인덱스가 설정됨\
원하는 컬럼즈를 설정하려면 매개변수 columns에 설정하고 싶은 배열을 입력함

````python
df = pd.DataFrame(data1, index=['김판다', '강승주', '송중기', '권보아'],
                  columns=['국어', '영어', '수학'])
df
````

데이터 프레임은 딕셔너리를 사용하여 생성할 수 있음\
이때는 밸류즈, 인덱스, 컬럼즈가 모두 포함된 2차원 딕셔너리를 생성하여 DataFrame 함수에 입력함\
컬럼즈가 상위 딕셔너리의 키(key)가 되고, 하위 딕셔너리에 인덱스와 데이터가 키와 밸류로 입력되는 2차원 구조

### 3.1.4 매개변수와 인수, 기본값
$f(x, y) = 2x + y $ 일 때 $f(2, 1)$의 값은?\
위 문제는 어렵지 않게 풀 것인데, x의 자리에 2를 넣고 y의 자리에 1을 넣어서 정답 5를 도출할 것임\
함수 f에서 x와 y는 함수의 입력값을 나타내는 변수로, 함수를 정의할 때 값이 입력될 자리를 만들어둔 것이며 이 x,y를 매개변수라 함\
$f(2, 1)$에서 2와 1은 함수의 입력값으로 매개변수 x와 y에 전달되는 값이며 매개변수에 전달되는 입력값이 인수임\
위 예시에서 매개변수 x의 인수는 2, y의 인수는 1임

코딩에서 매개변수와 인수를 동시에 표기할 때가 많은 이유는 인수만 입력할 때는 매개변수를 입력하지 않아도 되어 간편하지만, 순서를 엄격히 따라 모든 인수를 입력해야 하는 단점이 존재하기 때문\
매개변수를 인수와 함께 입력하면 순서에 구애받지 않고, 일부의 매개변수에만 인수를 입력해서 함수를 사용할 수 있음\
내가 인수를 입력하지 않은 매개변수에도 인수는 전달되어야 하기에 함수를 생성할 때 인수를 입력하지 않아도 매개변수에 전달될 값을 정의함\
별도의 인수를 입력하지 않아도 전달되는 인수를 기본값이라고 함\
위의 데이터프레임 예시에서 매개변수 index와 columns에 인수를 입력하지 않아 기본값으로 RangeIndex가 설정된 이유임\
매개변수의 순서에 맞게 인수를 입력하면 매개변수를 생략할 수 있는데, 너무 많이 생략하면 코드 가독성이 떨어지는 경우가 발생하므로 첫 번째 매개변수를 처음 입력할 때 등 몇몇 경우에만 매개변수를 생략함

## 3.4 데이터 프레임 저장하기
코드로 데이터 프레임을 생성하거나 파일에서 불러온 데이터 프레임을 기반으로 새로운 데이터 프레임을 생성한 후, 그 겨로가물을 다른 형태로 변환해야 할 때가 있는데, 데이터 프레임을 딕셔너리로 변환하거나 파일로 저장하는 방법을 살펴 봄

### 3.4.1 딕셔너리로 변환하기(to_dict)
데이터 프레임은 파이썬의 판다스에서만 사용 가능한 데이터 구조이기에 다른 프로그래밍 언어로 결과 데이터를 활용하고자 할 때나 파이썬 내에서 판다스가 아닌 다른 라이브러리로 데이터를 다룰 때도 주로 딕셔너리로 변환함

데이터 프레임은 to_dict 함수로 딕셔너리로 변환함

````python
df.to_dict()
````

### 3.4.2 파일로 저장하기(to_excel, to_csv)
데이터 프레임인 변수 ㅇf를 파일로 저장할 것인데 먼저 엑셀 파일로 저장할 때 to_excel 메서드를 사용함

````python
df.to_excel('1.xlsx')
````

위처럼 저장하면 인덱스도 함께 저장되는데 RangeIndex는 인덱스가 특별한 데이터를 갖지 않으므로 인덱스는 제외하고 파일로 저장해야함\
to_excel 메서드의 매개변수 index에 False를 입력하면 인덱스는 제외하고 파일로 저장됨

````python
df.to_excel('1.xlsx', index=False)
````

csv 파일로 저장하려면 to_csv 함수에 파일명을 인수로 입력함\
기본값으로 구분자는 쉼표로 지정되는데, 2차원 표 데이터를 쉼표로 열을 구분하고 줄 바꿈으로 행을 구분해 텍스트로 저장함\
인코딩 방식은 기본값으로 UTF-8로 지정됨

````python
df.to_csv('2.csv', index=False)
````

### 3.4.3 함수와 메서드, 속성
데이터 프레임을 저장하고자 사용한 to_dict, to_excel, to_csv 함수는 모두 메서드인데 함수와 메서드의 차이에 대해 살펴보자

메서드는 함수(function)의 일종이라서 매개변수에 인수를 전달받아 값을 반환함\
그러나 함수는 독립적으로 호출할 수 있는 반면, 메서드는 클래스 내부에 정의되어 객체에 종속되는 함수이기에 특정 객체에 종속되어 호출됨\
함수는 객체와 관계없이 호출할 수 있지만, 메서드는 객체를 먼저 불러온 후 호출해야 하므로 메서드는 객체 뒤에 붙여 사용함

1. pd.read_excel('파일경로명') : 함수, 판다스 함수는 앞에 pd.를 붙여 사용하며, 인수를 입력받아 결과를 반환함
2. df.to_dict() : 메서드, 객체에 종속되어 호출되므로 객체(df)를 불러온 뒤 메서드를 객체 뒤에 붙여 사용함
3. df.index : 속성, 속성은 클래스 내부에서 정의된 변수로 객체 뒤에 붙여 사용하고 결과를 반환하지만 함수가 아니라 인수를 받지 않음

속성과 메서드도 혼동하기 쉬운데, 속성은 메서드처럼 객체 뒤엥 붙여 사용하지만 클래스 내부에 정의된 변수를 반환하므로 인수가 입력되지 않음\
대부분의 함수를 메서드 형태로 사용할 것인데, 판다스에서 함수로 표기하더라도 대부분은 메서드를 지칭함\
함수보다 메서드를 주로 사용하는 이유는 판다스는 하나의 객체에 여러 가지 함수를 연속해서 사용하는 작업이 많기 때문임\
그래서 함수를 여러 번 사용하기보다 메서드를 연속해서 사용하여(chain method) 코드를 작성하면, 가독성이 높아지고 유지보수가 쉬움\
```py
연속함수 pd.funcC(pd.funcB(pd.funcA(df)))
연속 메서드 df.funcA().funcB().funcC() # funcA, funcB, funcC의 순차적 사용을 한눈에 파악하기 쉬움
```

### 4.1.5 키 인덱싱과 슬라이싱(loc 인덱서)
만약 A 학새으이 국어 성적이 궁금하다면?\
이런 작업은 대괄호 인덱싱으로는 두 번의 대괄호 인덱싱으로 추출해야 함\
대괄호 인덱싱은 2차원 구조인 데이터 프레임의 인덱싱을 완전히 만족시킬 수 없기에 간단한 인덱싱에 사용됨

대괄호 인덱싱과 달리 loc 인덱서와 iloc 인덱서는 모두 행과 열을 동시에 인덱싱할 수 있음\
loc 인덱서는 (key) location의 약자로서, 키를 기반으로 인덱싱을 수행하므로 열 이름을 입력함\
iloc 인덱서는 integer location의 약자로, 위치를 정수로 부여한 로케이션을 기반으로 인덱싱을 수행하므로 로케이션을 입력함

````python
df2 = df.copy()
df2.loc['A', '국어']
````

loc 인덱서는 행과 열을 동시에 인덱싱할 수 있기에 셀의 데이터를 추출할 수 있음\
loc 인덱서 역시 복수 인덱싱이 가능한데, 복수일 때는 이름들을 리스트로 입력해야 함

````python
df2.loc[['A', 'C', 'D'], ['국어', '수학']]
````

loc 인덱서도 슬라이싱이 가능함

````python
df2.loc['B':'D', '영어':'과학']
````

슬라이싱할 때 시작점이 데이터 프레임의 처음이거나, 끝 지점이 데이터 프레임의 마지막이라면 생략할 수 있음

````python
df2.loc['A':'D', '수학':'불어']
````

````python
df2.loc[:'D', '수학':]
````

데이터 프레임의 처음부터 끝까지 슬라이싱할 때, 시작점과 끝 지점을 모두 생략할 수 있는데 행의 슬라이싱에서는 콜론만 입력하면 됨

````python
df2.loc['A':'E', '수학':'과학']
````

````python
df2.loc[:,'수학':'과학']
````

모든 열을 가져올 때는 열의 슬라이싱 전체를 생략할 수 있으며 쉼표를 사용하지 않고 행의 슬라이싱만 입력함

````python
df2.loc[:'D',:]
````

````python
df2.loc[:'D']
````

하지만 행을 전부 가져올 때는 행의 슬라이싱은 생략될 수 없음

````python
df2.loc[:, '수학':'과학']
````

````python
df2.loc['수학':'과학']
````

행은 인덱싱을 수행하지만 열은 슬라이싱을 수행할 수 있고, 행은 슬라이싱을 수행하지만 열은 인덱싱을 수행할 수 있음

````python
df2.loc[:'D',['과학','수학']]
````

### 4.1.6 loc 인덱서로 행과 열 생성하기
열 생성은 대괄호 인덱서를 사용하는 것이 간편하므로 일반적으로 열의 생성은 대괄호 인덱서를 사용하는 것이 바람직함\
일부에만 값을 부여하는 열을 생성할 때는 loc 인덱서를 사용하는 것이 효과적임

````python
df2.loc[['B','D'],'실험'] = 80
df2
````

주로 빅데이터처럼 대규모 데이터를 다루므로 행 이름을 일일이 입력하여 특정 행에만 값을 부여하는 열을 생성하지는 않지만 조건에 따라 값을 다르게 부여하는 열을 생성할 때 이 방식을 사용함\
열 생성 외에 셀의 데이터도 수정할 수 있음

````python
df2.loc['A','과학']= 70
df2
````

loc 인덱서로는 행도 생성할 수 있는데, 존재하지 않는 행을 인덱싱해 값을 부여하면 새로운 행이 생성되며 데이터 프레임의 마지막에 위치함

````python
df2.loc['F'] = 60
df2
````

### 4.1.7 로케이션 인덱싱과 슬라이싱(iloc 인덱서)
데이터 프레임은 로케이션으로 인덱싱과 슬라이싱이 가능하며 이때는 iloc 인덱서를 활용함\
로케이션은 가장 처음의 원소에 0을 부여하고, 그 뒤로 1씩 순차적으로 증가하는 정수를 부여함

````python
df3 = df.copy()
df3.iloc[0,0]
````

iloc 인덱서 역시 복수의 행과 열을 인덱싱하는 것도 가능한데, 복수의 행과 열을 인덱싱할 때는 로케이션을 리스트로 묶어 입력함

````python
df3.iloc[[0, 2, 3], [0, 2]]
````

iloc 인덱서도 슬라이싱을 수행하며 시작점과 끝 지점의 로케이션을 콜론으로 구분해서 입력함

````python
df3.iloc[1:3, 1:4]
````

loc 인덱서는 양쪽 경계를 모두 포함하지만, iloc 인덱서는 시작점은 포함하고 끝지점은 포함하지 않음\
이는 리스트의 슬라이싱과 동일한 방식이며, 파이썬의 로케이션 슬라이싱이 가지는 공통적인 특징임\
iloc 인덱서도 슬라이싱할 때 시작점이 데이터 프레임의 처음이거나, 끝 지점이 데이터 프레임의 마지막이면 생략할 수 있음

````python
df3.iloc[:4, 2:]
````

행을 모두 가져올 때는 시작점과 끝 지점이 모두 생략되어 콜론만으로 표현할 수 있음\
열을 모두 가져올 때는 열의 슬라이싱을 전체적으로 생략할 수 있음

````python
df3.iloc[0]
````

````python
df3.iloc[-1]
````

또한 인덱싱과 슬라이싱은 행과 열에 각각 적용되며, 인덱싱이 순서에 구애받지 않는다는 점은 iloc 인덱서도 갖는 특징임

````python
df3.iloc[1:4, [-2, 3, -2]]
````

### 4.2.1 행과 열 삭제하기(drop)
판다스 drop\
데이터 프레임의 행이나 열을 삭제하는 함수
```py
drop 함수의 주요 매개변수와 인수, 기본 값
df.drop(labels, axis=0, level=None)
```
* labels : 삭제할 행 또는 열을 지정하며 복수의 행이나 열을 지정하려면 리스트로 묶어서 입력함
* axis : 행을 삭제할지 열을 삭제할지 지정하며, 기본값은 0이고 행을 삭제함
* level : 멀티 인덱스일 때 삭제할 레벨을 지정함

````python
import pandas as pd
data1 = [[95, 65, 82], [84, 83, 76], [91, 88, 69], [91, 67, 90]]
idx = ['A', 'B', 'C', 'D']
cols = columns=['국어', '영어', '수학']
df = pd.DataFrame(data1, index=idx, columns=cols)
df
````

````python
df.drop('영어', axis=1)
````

````python
df.drop(['영어', '수학'], axis=1)
````

````python
data2 = {'이름': ['김판다', '강승주', '권보아', '임재범'],
         '국어점수': [88, 81, 89, 79], '영어점수': [68, 82, 84, 80],
         '수학점수': [75, 80, 84, 89], '과학점수': [75.5, 62, 82, 89],
         '성별': ['남', '여', '여', '남'], '반': ['A', 'B', 'A', 'B'],
         '전형': ['수시', '수시', '정시', '정시']}
df1 = pd.DataFrame(data2)
df1
````

````python
df1.iloc[:, 1:7].drop('과학점수', axis=1)
````

2. 서로 다른 함수나 기법으로 키와 로케이션을 각각 적용할 때(loc인덱서, iloc인덱서)는 상황에 따라 편한 쪽을 선택함
3. 키만 적용 가능할 때(drop 함수)에는 drop 함수는 삭제할 키만 입력할 수 있고 로케이션인 0을 사용할 수는 없기에 df.columns를 활용하면 로케이션으로 열 이름을 반환할 수 있다는 것을 이용하여 df.drop(df.columns[0], axis=1)과 같은 코드로 로케이션을 활용하여 첫 번째 열을 삭제함

### 4.2.2 열 이름으로 필터링(filter)

dㅕㄹ 이름으로 데이터 프레임을 필터링하는 함수

filter함수의 주요 매개변수와 인수, 기본값

### 4.2.4 판다스의 함수와 원본 변경하기

판다스 함수는 원본을 변경하지 않아 연속 메서드를 사용할 수 있으며, 별도의 객체를 생성해 연속 메서드로 함수를 적용함

## 4.3 인덱스와 컬럼즈를 다루는 함수들
인덱싱은 데이터 프레임에서 인덱스와 컬럼즈를 기반으로 원하는 데이터만 가져오는 필수적인 기능\
filter 함수도 컬럼즈를 기반으로 작동하여 간결하고 강력한 기능을 수행함\
이번 절에서 학습하는 인덱스와 컬럼즈를 다루는 함수들은 대부분 인덱스 클래스에 적용하는 것이 아니라 데이터 프레임에 적용해 인덱스와 컬럼즈를 변경하는 함수들임

### 4.3.1 인덱스 설정하기(set_index)
판다스 set_index\
데이터 프레임의 인덱스를 설정하는 함수(컬럼즈를 설정할 수는 없음)
```py
set_index 함수의 주요 매개변수와 인수, 기본값
df.set_index(keys, axis=0)
```
* keys : 인덱스로 설정될 열을 지정함

set_index는 일반 열을 인덱스로 설정하는 함수\
열을 인덱스로 설정할 때만 적용이 가능한 함수이며 행을 컬럼즈로 설정할 수는 없음

````python
import pandas as pd
data = [['송중기', 70, 'A'], ['강승주', 80, 'A'],
        ['김판다', 93, 'B'], ['이예준', 92, 'B']]
df = pd.DataFrame(data, columns=['이름', '점수', '반'])
df1 = df.set_index('이름')
df2 = df.copy() # df를 복제한 df2
df
````

````python
df.set_index('이름')
````

기존의 열 이름인 '이름'은 인덱스명이 됨\
인덱스명을 변경하거나 제거하기를 원할 때는 rename_axis 함수를 사용함\
인덱스명은 단일 인덱스를 다룰 때는 큰 의미가 없지만 멀티 인덱스를 다룰 때는 인덱스명이 있으면 편리함

````python
df.set_index(['반', '이름']) # df.loc[('A', '송중기'), '점수']의 값은 70
````

### 4.3.2 인덱스 리셋하기(reset_index)
판다스 reset_index\
인덱스에서 데이터를 제거하는 함수(컬럼즈를 대상으로는 사용할 수 없음)\
인덱스에서만 제거해 데이터 프레임의 열이 되거나, 데이터를 완전히 삭제할 수 있고 인덱스가 모두 제거되면 새로운 RangeIndex가 생성됨
```py
reset_index 함수의 주요 매개변수와 인수, 기본값
df.reset_index(level=None, drop=False)
```
* level : 멀티 인덱스에서 리셋할 인덱스의 레벨을 지정함
* drop : 인덱스에서만 제거할지 완전히 데이터를 삭제할지 지정하는데 기본값은 인덱스의 데이터가 열이 됨

````python
df1.reset_index()
````

````python
df1.reset_index(drop=True)
````

### 4.3.3 인덱스와 컬럼즈 변경하기(배정)

### 5.4.1 집계함수
집계 함수(ㅁggregation function)는 군집의 데이터를 요약한 통계를 제공하는 함수\
합(sum), 평균(mean), 표준편차(std), 최댓값(max), 최솟값(min), 중간값(median) 등 대부분의 통계 함수가 여기에 속함

````python
df1.mean()
````

열마다 열의 평균을 반환하고 결과가 모여서 시리즈가 됨\
집계 함수를 데이터 프레임에 적용하면 차원이 축소되어 시리즈를 반환함\
이렇게 그룹에 적용하여 값을 반환해 차원을 축소하는 함수가 집계 함수이며, 집계 함수는 시리즈에도 적용이 가능함(단일 값을 반환함)\
결과로 각 열의 평균이 반환되었는데 만약 각 행의 평균을 반환받고 싶다면 축을 axis=1로 지정하면 됨

이상치는 데이터의 정확성이나 신뢰성에 영향을 미치므로 이를 식별하여 처리하는 것이 중요함\
이상치를 식별하고 대체하거나 제거할 때 여러 가지 방법을 사용하므로 하나의 함수로 모든 이상치를 처리할 수는 없음\
여기서는 다양한 이상치 처리 방법 중 가장 간단한 방법인 상한선과 하한선의 임곗값으로 값을 대체하는 clip 함수를 학습할 것임

판다스 clip\
상한선이나 하한선으로 임곗값을 적용해 이상치를 처리하는 함수
```py
df.clip(lower=None, upper=None)
```
* lower : 최솟값으로 적용할 임곗값을 입력하며 배열을 입력하면 열마다 다르게 적용함
* upper : 최댓값으로 적용할 임곗값을 입력하며 배열을 입력하면 열마다 다르게 적용함

````python
import pandas as pd
data1 = [[89, 4, 74], [29, 46, 83], [40, 19, 60], [29, 91, 76]]
df1 = pd.DataFrame(data1, index=list('ABCD'), columns=['국어', '영어', '수학'])
df1
````

````python
df1.clip(lower=20, upper=80)
````

clip 함수에 배열을 입력하면 열마다 다른 임곗값이 지정되는데 clip 함수는 간편함이 장점임\
다만 clip 함수만으로는 복잡한 이상치 처리에 한계가 있으므로, 불리언 마스킹을 비롯한 여러 가지 판다스 기법과 함수를 활용함

### 7.1.1 데이터 프레임 연결하기(concat)
판다스 concat\
데이터 프레임을 연결하는 함수
```py
pd.concat(objs, axis=0, join='outer')
```
* objs : 연결할 시리즈나 데이터 프레임들을 리스트로 묶어 입력함
* axis : 연결할 축 방향을 지정하며 기본값은 세로 방향으로 연결함
* join : 결합 방식을 지정함

````python
import pandas as pd
data1 = [[80, 69, 83, 98], [71, 90, 69, 66], [74, 72, 72, 95]]
data2 = [[68, 70, 84, 70], [65, 91, 90, 66], [78, 94, 96, 64]]
data3 = [[65, 82], [85, 60], [75, 78]]
data4 = [[79, 99, 95, 90], [99, 76, 81, 97], [95, 80, 62, 84]]
col1 = ['국어', '수학', '영어', '과학']
col2 = ['사회', '일어']
col3 = ['국어', '수학', '영어', '사회']

df1 = pd.DataFrame(data1, index=list('ABC'), columns=col1)
df2 = pd.DataFrame(data2, index=list('DEF'), columns=col1)
df3 = pd.DataFrame(data3, index=list('ABC'), columns=col2)
df4 = df2[['영어', '과학', '수학', '국어']]
df5 = pd.DataFrame(data4, index=list('GHI'), columns=col3)
````

````python
df1
````

````python
df2
````

````python
pd.concat([df1, df2])
````

````python
pd.concat([df1, df4])
````

### 7.1.2 외부 조인과 내부 조인
세로 방향으로 연결할 때 두 데이터 프레임이 보유한 열이 다르면 어떻게 될까?

이러한 결합 방식을 외부조인(outer join)이라고 하며, concat 함수의 결합 방식의 기본값은 외부조인\
열의 교집합을 기준으로 연결되는 결합 방식은 내부조인(inner join)이며, join='inner'를 입력하면 됨

## 7.2 데이터 프레임 병합하기
엑셀의 vlookup 방식으로 두 데이터 프레임을 결합하는 것을 병합이라고 하며 판다스는 merge 함수로 병합을 수행함

### 7.2.1 엑셀의 vlookup 방식으로 병합하기(merge)
판다스 merge\
데이터 베이스의 조인을 수행하는 함수이며 주로 엑셀의 vlookup 방식의 병합을 수행함
```py
merge 함수의 주요 매개변수와 인수, 기본값
df.merge(right, how='inner', on=None)
```
* right : 병합할 객체를 지정함
* how 병합 방식을 지정하며 'left'가 vlookup 방식임
* on : 병합의 기준이 되는 열을 지정하며 기본값은 양측에 공통으로 존재하는 열 이름 기준으로 병합함

````python
import pandas as pd
data1 = {'이름': ['김판다', '조민영', '강승주', '최진환', '박연준'],
         '제품': ['아이스티', '카페라떼', '카페라떼', '아이스티', '녹차']}
data2 = {'제품': ['카페라떼', '아이스티', '녹차'], '가격': [4500, 4600, 4800]}
data3 = {'상품': ['카페라떼', '아이스티', '녹차'], '가격': [4500, 4600, 4800]}
data4 = {'가격': [4500, 4600], '제품': ['카페라떼', '아이스티'],
         '분류': ['커피', '음료']}
data5 = {'제품': ['카페라떼', '아이스티', '녹차', '녹차'],
         '가격': [4500, 4600, 4800, 3500],
         '비고': ['정상가', '정상가', '정상가', '할인가']}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
df3 = pd.DataFrame(data3)
df4 = pd.DataFrame(data4)
df5 = pd.DataFrame(data5)
````

````python
df1
````

````python
df2
````

````python
df1.merge(df2, how='left')
````

병합의 기준이 될 열을 매개변수 on에 지정해야 하는데, 기본값으로 양측 데이터 프레임에 공통으로 존재하는 열 이름을 기준으로 병합을 수행함\
df1과 df2 양측에 모두 제품 열을 가져 아무것도 입력하지 않아도 기본값으로 제품 열을 기준으로 병합함\
병합의 기준이 되는 열 이름이 양측 데이터 프레임에 다르다면 어떻게 해야하는가?\
여러 가지 방법이 가능하지만 rename 함수로 양측 데이터 프레임의 열 이름을 일치시키는 것이 가장 간편함

````python
df1.merge(df3.rename({'상품':'제품'},axis=1), how='left')
````

물론 merge 함수는 left_on, right_on 등의 매개변수로 병합할 기준이 될 열을 각각 다르게 지정할 수 있음\
merge 함수는 vlookup 함수와 달리 열의 순서와 무관하며 한 번에 여러 개의 열도 병합할 수 있고 복수의 행도 한 번에 병합함

## 7.3 업데이트
연결과 병합 외에도 기존의 데이터 프레임에 새로운 데이터 프레임을 업데이트하는 것이 가능함\
판다스는 update 함수와 combine_first 함수를 사용하여 업데이트를 수행함

### 7.3.1 데이터 프레임 업데이트하기(update)
판다스 update\
입력된 데이터 프레임의 null이 아닌 값으로 기존 데이터 프레임을 수정하는 함수
```py
update함수의 주요 매개변수와 인수, 기본값
df.update(other=None)
```
* other : 업데이트할 새로운 데이터 프레임을 입력함

update 함수는 기존 데이터 프레임을 새로운 데이터 프레임으로 업데이트하는 함수\
새로운 데이터 프레임의 null은 업데이트 대상이 아님\
결과는 기존 데이터 프레임을 직접 수정하는데 판다스에서 함수 적용 결과가 원본을 변경하는 몇 안 되는 함수

````python
import pandas as pd
data1 = {'국어': {'A': 25, 'B': 72, 'C': 13, 'D': 78, 'E': 25},
         '영어': {'A': 45, 'B': 94, 'C': 92, 'D': 70, 'E': 18}}
data2 = {'국어': {'A': 64, 'E': 75, 'C': 94},
         '영어': {'A': 76, 'E': 85, 'C': float('nan')}}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
````

````python
df1
````

````python
df2
````

````python
df1.update(df2)
````

````python
df1
````

인덱스와 컬럼즈에 맞추어 df2의 데이터가 df1에 업데이트됨\
df2에서 C행, 영어 열의 데이터는 NaN이므로 업데이트 되지 않고 df1의 점수인 92가 유지됨

### 7.3.2 데이터 프레임 업데이트하기(combine_first)
판다스 combine_first\
두 객체에서 null이 아닌 첫 번째 값을 반환하는 함수
```py
combine_first 함수의 주요 매개변수와 인수, 기본값
df.combine_first(other=None)
```
* other : 두 번째 객체를 입력함

판다스에서는 원본을 변경하는 것을 기피하는 데, 코드가 원본 데이터에 직접 접근하면 코드의 안정성이 저하되고, 연속 메서드도 쓸 수 없고, 여러 번 실행할 때 기존 변수부터 다시 실행해야 해서 번거롭기 때문임

combine_first 함수는 두 객체에서 null이 아닌 첫 번째 값을 반환하는 함수이기에 update 함수와 비슷한 기능을 수행하며 원본을 변경하지 않는 장점이 존재함

````python
df2
````

````python
df2.combine_first(df1)
````

## 7.4 범위로 병합하기
정확한 값이 아닌 범위의 값으로 병합해야 할 때 판다스에서는 merge_asof 함수를 사용함

### 7.4.1 범위로 병합하기(merge_asof)
판다스 merge_asof\
범위를 기준으로 병합을 수행하는 함수이며 엑셀의 index와 match 조합과 비슷함
```py
merge_asof 함수의 주요 매개변수와 인수, 기본값
pd.merge_asof(left, right, on=None, by=None)
```
* left, right : 병합할 객체를 입력함
* on : 병합의 기준이 되는 열을 지정하며 반드시 오름차순으로 정렬되어야 함
* by : 그룹화해서 병합할 때 그룹의 기준을 지정함
* 그외 allow_exact_matches, direction 등의 매개변수가 존재함

````python
import pandas as pd
data1 = {'이름': ['라', '나', '다', '가', '마'],
         '점수': [60, 72, 80, 88, 95]}
data2 = {'점수': [0, 70, 80, 90], '학점': ['F', 'C', 'B', 'A']}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
````

````python
df1
````

````python
df2
````

````python
pd.merge_asof(df1, df2, on='점수')
````
