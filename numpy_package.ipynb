# 넘파이 패키지
파이썬이 대량 데이터를 대상으로 정교한 수학 연산을 수행하는 패키지\
Numerical Python의 준말인 넘파이는 파이썬에서 산술 계산을 위한 가장 중요한 필수 패키지 중 하나\
과학 계산을 위한 대부분의 패키지는 넘파이의 배열 객체를 데이터 교환을 위한 공통 언어처럼 사용함

넘파이에서 제공하는 기능
* 빠른 배열 계산과 유연한 브로드캐스팅 기능을 제공하는 효율적인 다차원 배열인 ndarray
* 반복문을 작성할 필요 없이 전체 데이터 배열을 빠르게 계산하는 표준 수학 함수
* 배열 데이터를 디스크에 쓰거나 읽을 수 있는 도구와 메모리에 적재된 파일을 다루는 도구
* 선형대수, 난수 생성기, 푸리에 변환 기능
* C, C++, 포트란으로 작성된 코드를 넘파이와 연결하는 C API

넘파이는 포괄적이고 잘 문서화된 C API를 제공하므로 저수준 언어로 작성된 외부 라이브러리에 데이터를 전달하거나 외부 라이브러리에서 데이터를 넘파이 배열로 파이썬에 전달하기가 용이함\
이러한 기능 덕분에 파이썬으로 레거시 C, C++, 포트란 코드를 감싸서 동적이면서도 접근이 가능한 인터페이스를 제공할 수 있음

넘파이 자체는 모델링이나 과학 계산을 위한 기능을 제공하지 않으므로 먼저 넘파이 배열과 배열 기반 연산에 대해 이해한 후 판다스 같은 배열 기반 도구를 사용하면 훨씬 더 효율적임

대부분의 데이터 분석 애플리케이션에서 중요한 기능은 다음과 같음
* 데이터 먼징(munging)(데이터 랭글링), 정제, 부분집합, 필터링, 변형, 다른 여러 종류의 연산을 빠르게 수행하는 배열 기반 작업
* 정렬, 유일 원소 찾기, 집합 연산 같은 일반적인 배열 처리 알고리즘
* 효과적인 기술 통계와 데이터의 수집, 요약
* 다양한 종류의 데이터를 병합하고 엮기 위한 데이터 정렬과 데이터 간의 관계 조작
* if-elif-else를 사용하는 반복문 대신 배열 표현식으로 조건부 로직 표현
* 수집, 변형, 함수 적용 같은 그룹별 데이터 조작

넘파이는 대용량 데이터 배열을 효율적으로 다루도록 설계되었기에 파이썬 산술 계산 영역에서 중요한 위치
* 넘파이는 내부적으로 데이터를 다른 내장 파이썬 객체와는 구분된 연속된 메모리 블록에 저장함\
넘파이의 각종 알고리즘은 모두 C로 작성되어 타입 검사나 다른 오버헤드 없이 메모리를 직접 조작할 수 있음\
또한 넘파이 배열은 내장 파이썬의 연속된 자료형보다 훨씬 더 적은 메모리를 사용함
* 넘파이 연산은 큰 배열을 다루는 경우 속도가 느려질 수 있는 파이썬 반복문을 사용하지 않고, 전체 배열에 대한 복잡한 계산을 수행함\
넘파이는 C 기반 알고리즘을 사용해 일반 파이썬 코드 실행에서 발생하는 오버헤드를 줄임으로써 더 빠르게 작동함

넘파이의 핵심 기능은 ndarray라고 하는 N차원의 배열 객체\
ndarray는 파이썬에서 사용할 수 있는 대규모 데이터셋을 담을 수 있는 빠르고 유연한 자료구조\
배열을 사용하면 스칼라 원소 간의 연산에 사용하는 문법과 비슷한 방식을 사용해 전체 데이터 블록에 수학적 연산을 수행할 수 있음

## 1. array, numpy, matplotlib 패키지 개요
### 1.1 array 패키지
array 패키지는 numpy의 기본적인 기능을 많이 가지고 있으며 다른 프로그램에서 생성한 연속적인 데이터 더미와도 인터페이스할 수 있게 해주기 때문에 유용하게 사용 가능함

### 1.2 numpy 패키지

numpy.random 패키지는 numpy 패키지의 일부이며, random 패키지와 동일한 기능을 제공하지만, numpy 배열과 함께 사용하도록 최적화됨

### 1.4 matplotlib 패키지

## 2. array 패키지란?
array 패키지는 파이썬 리스트 대신 C와 C++에서 다루는 배열의 특징을 가짐\
`배열`은 색인 번호로 메모리 속에 있는 각 항목을 참조할 수 있는 일련의 집합\
리스트와 달리 배열은 고정 길이 데이터라고 가정함\
데이터의 모든 항목이 메모리상에 연속적으로 서로 옆에 위치한다는 의미


````python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False

fig, axes = plt.subplots(1, 2, figsize=(15, 7))
fig.patch.set_facecolor('#1e1e2e')

for ax in axes:
    ax.set_facecolor('#1e1e2e')
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 10)
    ax.axis('off')

# ── 왼쪽: Python List ──────────────────────────────────────
ax = axes[0]
ax.set_title('Python List', color='white', fontsize=14, fontweight='bold', pad=10)

# List Object 박스
list_box = mpatches.FancyBboxPatch((0.4, 5.5), 3.2, 4.0,
    boxstyle="round,pad=0.1", linewidth=2,
    edgecolor='#89b4fa', facecolor='#313244')
ax.add_patch(list_box)
ax.text(2.0, 9.7, 'List Object', color='#89b4fa', fontsize=10,
        ha='center', fontweight='bold')

# 포인터 슬롯 3개
slot_labels = ['ref[0]  ──►', 'ref[1]  ──►', 'ref[2]  ──►']
slot_ys = [8.7, 7.5, 6.3]
for label, sy in zip(slot_labels, slot_ys):
    slot = mpatches.FancyBboxPatch((0.6, sy - 0.3), 2.8, 0.6,
        boxstyle="round,pad=0.05", linewidth=1,
        edgecolor='#6c7086', facecolor='#45475a')
    ax.add_patch(slot)
    ax.text(2.0, sy, label, color='#cdd6f4', fontsize=9, ha='center', va='center')

# 타입 객체 박스 (중간 단계)
type_labels = ['int 객체', 'int 객체', 'list 객체']
type_xs = [5.0, 5.0, 5.0]
type_ys = [8.7, 7.5, 6.3]
type_colors = ['#fab387', '#fab387', '#cba6f7']

for (tx, ty), tlabel, tcolor in zip(zip(type_xs, type_ys), type_labels, type_colors):
    tbox = mpatches.FancyBboxPatch((tx - 0.7, ty - 0.3), 1.4, 0.6,
        boxstyle="round,pad=0.05", linewidth=1.5,
        edgecolor=tcolor, facecolor='#313244')
    ax.add_patch(tbox)
    ax.text(tx, ty, tlabel, color=tcolor, fontsize=8.5, ha='center', va='center')
    # List Object → 타입 객체 화살표
    ax.annotate('', xy=(tx - 0.7, ty),
                xytext=(3.4, ty),
                arrowprops=dict(arrowstyle='->', color='#f38ba8', lw=1.5))

# 실제 값 객체들
val_data = [
    (8.0, 8.7, '10', '#a6e3a1'),
    (8.0, 7.5, '20', '#a6e3a1'),
    (8.0, 6.8, '5',  '#89dceb'),
    (8.0, 6.3, '12', '#89dceb'),
    (8.0, 5.8, '7',  '#89dceb'),
    (8.0, 5.3, '3',  '#89dceb'),
]
for vx, vy, vval, vcolor in val_data:
    circle = plt.Circle((vx, vy), 0.4, color=vcolor, linewidth=1.5, fill=True, zorder=3)
    ax.add_patch(circle)
    ax.text(vx, vy, vval, color='#1e1e2e', fontsize=9,
            ha='center', va='center', fontweight='bold', zorder=4)

# 타입 객체 → 값 화살표
for (tx, ty), (vx, vy) in zip(zip(type_xs[:2], type_ys[:2]), [(8.0, 8.7), (8.0, 7.5)]):
    ax.annotate('', xy=(vx - 0.4, vy), xytext=(tx + 0.7, ty),
                arrowprops=dict(arrowstyle='->', color='#f38ba8', lw=1.5))

# list 객체 → [5,12,7,3] 화살표들
list_vals_ys = [6.8, 6.3, 5.8, 5.3]
for lvy in list_vals_ys:
    ax.annotate('', xy=(8.0 - 0.4, lvy), xytext=(5.7, 6.3),
                arrowprops=dict(arrowstyle='->', color='#cba6f7', lw=1.2,
                                connectionstyle='arc3,rad=0.0'))

ax.text(4.0, 4.6, '참조(ref)를 통해 각 객체에 접근', color='#f38ba8',
        fontsize=8, ha='center', style='italic')

# ── 오른쪽: NumPy Array ────────────────────────────────────
ax = axes[1]
ax.set_title('NumPy Array (ndarray)', color='white', fontsize=14,
             fontweight='bold', pad=10)

# ndarray 객체 박스
arr_box = mpatches.FancyBboxPatch((0.4, 6.5), 3.2, 3.0,
    boxstyle="round,pad=0.1", linewidth=2,
    edgecolor='#89b4fa', facecolor='#313244')
ax.add_patch(arr_box)
ax.text(2.0, 9.7, 'ndarray Object', color='#89b4fa', fontsize=10,
        ha='center', fontweight='bold')

meta_lines = ['dtype: int64', 'shape: (3,)', 'data ptr']
meta_ys = [9.0, 8.3, 7.6]
for m, my in zip(meta_lines, meta_ys):
    ax.text(2.0, my, m, color='#cdd6f4', fontsize=9, ha='center')

# data ptr → 연속 블록 화살표
ax.annotate('', xy=(4.5, 7.0),
            xytext=(3.6, 7.6),
            arrowprops=dict(arrowstyle='->', color='#a6e3a1', lw=2.0))

# 연속 메모리 블록
block_x = 4.5
block_y = 6.3
cell_w, cell_h = 1.5, 1.2
values = [10, 20, 30]
colors = ['#89dceb', '#89dceb', '#89dceb']
for i, (v, c) in enumerate(zip(values, colors)):
    rect = mpatches.FancyBboxPatch((block_x + i * cell_w, block_y), cell_w, cell_h,
        boxstyle="square,pad=0.0", linewidth=2,
        edgecolor='#89b4fa', facecolor=c)
    ax.add_patch(rect)
    ax.text(block_x + i * cell_w + cell_w / 2, block_y + cell_h / 2,
            str(v), color='#1e1e2e', fontsize=14,
            ha='center', va='center', fontweight='bold')
    # 주소 표시
    ax.text(block_x + i * cell_w + cell_w / 2, block_y - 0.3,
            f'addr+{i*8}', color='#6c7086', fontsize=7.5, ha='center')

# 연속 강조 브레이스
ax.annotate('', xy=(block_x + 3 * cell_w, block_y - 0.7),
            xytext=(block_x, block_y - 0.7),
            arrowprops=dict(arrowstyle='<->', color='#a6e3a1', lw=1.5))
ax.text(block_x + 1.5 * cell_w, block_y - 1.1,
        '연속된 메모리 블록 (8 bytes 간격)',
        color='#a6e3a1', fontsize=8.5, ha='center')
ax.text(block_x + 1.5 * cell_w, block_y - 1.6,
        '(cache-friendly, 빠른 연산)',
        color='#fab387', fontsize=8.5, ha='center', style='italic')

plt.tight_layout()
plt.show()
````

파이썬 리스트에는 보통 눈에 띄지 않는 참조 번호가 포함되어 있음(C에서는 포인터가 될 것임)\
리스트 객체에는 실제 리스트 위치에 접근할 수 있는 참조가 있으며, 이 참조는 변경될 수도 있음\
한편 리스트 항목에 위치한 각 객체 역시 실제 데이터에 접근할 수 있는 참조이기 때문에 파이썬에서는 한 리스트 안에서도 여러 데이터 타입인 항목을 가질 수 있는 것임

위 그림에서 알 수 있듯이 배열은 설계가 더 간단한데, 배열 객체는 그 자체가 메모리 속 위치에 접근할 수 있는 참조일 뿐임\
실제 데이터가 그 위치에 있다는 의미임\
데이터가 이런 식으로 저장되기에 모든 항목은 길이가 일정해야 하며, 같은 타입이어야 함\
무작위 파이썬 정수(이론적으로 메모리 여러 바이트를 차지할 수도 있음)를 저장할 수는 없지만, 고정 길이의 정수를 저장할 수는 있음\
배열은 리스트보다 데이터를 더 압축하여 저장하지만 파이썬 리스트 색인은 매우 최적화되었기 때문에 배열의 색인보다 조금 느림

array 패키지를 사용할 때의 장점 중 하나는 이 장에서 설명하는 것처럼 배열이 저장되는 메모리 연속 블록에 데이터를 전달할 수 있다는 점\
array 패키지를 사용하려면 array 패키지를 탑재한 후 array.array를 호출하여 ㅁrray 객체를 대입하고 초기화하면 됨
```python
import array
a = array.array('h', [1,2,3])
```
여기에서 'h'를 첫 번째 인수로 사용하였는데, 데이터 타입을 지정하는 단일-문자 문자열임\
이 경우 16비트(2바이트) 정수(양수 혹은 음수 32k 범위)로 제한함\
range 함수를 사용하면 더 큰 배열을 만들 수 있음
```python
import array
a = array.array('h', range(1000))
```

## 4. numpy 소개
넘파이를 설치한 다음 단계는 다음 코드로 numpy를 탑재하는 것

````python
import numpy as np
````

numpy라는 이름이 길 수 있으므로 as np를 사용하여 numpy 대신 np라는 더 짧은 이름으로 패키지를 사용하라는 의미\
표준 numpy 루틴에 의해 생성된 데이터 타입을 ndarray라고 하며 'N-차원 배열'을 의미함\
numpy를 사용하는 이유는 무엇일까?\
1부터 100만까지 숫자를 더하는 문제를 생각해 보자

다음 코드는 파이썬 언어를 사용하여 숫자를 더하는 가장 효율적인 방법을 보여줌

````python
a_list = list(range(1, 1_000_001))
print(sum(a_list))
````

대개 기본적으로 제공하는 언어의 표준 방식을 사용하는 것도 나쁘지는 않음\
다음 코드는 같은 동작을 하는 numpy 기반 코드

````python
import numpy as np
a = np.arange(1, 1_000_001, dtype=np.int64)
print(sum(a))
````

두 코드의 실행 속도 차이를 측정하려면 성능 비교 테스트가 필요하며 time 패키지를 사용함

````python
import numpy as np
import time

def benchmarks(n):
    t1 = time.time()
    
    a_list = list((range(1, n+1)))
    tot = sum(a_list)
    
    t2 = time.time()
    print('Time taken by Python is ', t2-t1)
    
    t1 = time.time()
    
    a = np.arange(1, n+1)
    tot = np.sum(a)
    
    t2 = time.time()
    print('Time taken by Python is ', t2-t1)
````

````python
benchmarks(10_000_000)
````

## 5. numpy 배열 만들기
앞 절에서 큰 numpy 배열을 생성하는 한 가지 방법을 보여 주었음
```python
a = np.arange(1, n+1)
```
이 문장은 1로 시작하고 n까지의 범위를 생성하고 이 데이터를 사용하여 1차원 numpy 배열을 초기화함\
numpy 배열을 만들고 초기화하는 방법은 여러 가지가 있음
|넘파이 함수|생성 대상|
|:--|:--|
|arange|파이썬 range 함수와 유사한 구문을 사용하여 지정된 범위의 정수로 구성된 배열|
|linspace|지정된 범위 내에서 균일한 간격으로 배치된 값의 배열 <br> 이 함수는 부동소수점 값을 처리하므로, 원하는 경우 작은 분수 단위의 차이도 처리할 수 있음 <br>(기술적으로 정수를 수용할 수 있지만, 주로 부동소수점과 함께 사용하도록 설계되었음)|
|empty|초기화되지 않은 배열 <br> 값은 '무작위'이지만, 무작위 표본 추출에는 통계적으로 유효하지 않음|

### 5.1 array 함수(array로 변환)
numpy 배열을 생성하는 가장 간단한 방법은 파이썬 리스트나 튜플에서 array로 변환하는 방법임\
```python
numpy.array(data, dtype=None, order='K')
```
이 함수는 지정된 타입의 numpy 배열을 반환함\
dtype이 지정되지 않았거나 None으로 설정된 경우 이 함수는 모든 항목을 저장할 수 있을 만큼 큰 데이터 타입을 선택함\
(파이썬의 정수는 길이가 정해져 있지 않기 때문에 중요한 문제임)

order는 다차원 데이터가 어떻게 정렬되어야 할지 결정하는데, 기본값은 'K'이며, 어떤 값을 설정하든 소스 데이터를 보존함\
'C'는 행-우선 순서를 따르며(C언어에서 이 방식을 따름), 'F'는 열-우선 순서를 따름(포트란에서 이 방식을 따름)

다음 예시는 파이썬 리스트로 1차원 정수 배열을 초기화함

````python
import numpy as np
a = np.array([1,2,3])
a
````

다차원 파이썬 리스트(리스트를 항목으로 갖는 리스트)를 사용하는 것만으로 2차원 혹은 다차원 배열을 쉽게 만들 수 있음

만약 리스트를 항목으로 갖는 다차원 데이터를 '가지런하지 않게' 입력하더라도, 배열 변환은 최대한 일반적인 배열로 보정됨

### 5.3 linspace 함수
linspace 함수는 arange 함수와 비슷하지만 정수뿐만이 아니라 부동소수점도 다루며, 항목 간의 값 차이(step)도 자유롭게 설정가능\
이 함수는 특히 선을 따라 점 또는 값 집합을 제공하려고 할 때, 해당 값이 균등하게 간격을 두는 경우에 유용함\
arange 함수와 마찬가지로 linspace는 1차원 배열만 생성함
```python
numpy.linspace(begin, end, num=50, endpoint=True, dtype=None)
```
end의 값은 기본적으로 생성되는 값의 범위에 포함되며 endpoint 인수를 False로 설정하면 end 값은 포함되지 않음\
num 인수는 생성할 값 개수를 지정하며 가능한 범위 내에서 균일하게 간격을 두게 됨(지정하지 않으면 50을 기본값으로 설정함)\
dtype 인수는 각 항목의 데이터 타입을 지정하는데, 지정하지 않거나 None으로 설정하면 linspace 함수는 인수 나머지를 기반으로 데이터 타입을 추론하며, 일반적으로 float가 설정됨

````python
import numpy as np
a = np.linspace(0, 1.0, num=5)
a
````

````python
b = np.linspace(1, 5, num=5, dtype='int32')
b
````

### 5.4 empty 함수
empty 함수는 초기화하지 않은 numpy 배열을 생성함\
생성한 배열의 초깃값을 추후에 설정하는 경우라면, 시간을 절약하기 위해 두 번 중복으로 초기화하는 것을 막기 위한 용도로 empty 함수를 사용할 수 있음\
그러나 초기화하지 않은 객체를 사용하는 것은 위험한 작업이기 때문에 주의해야 함\
실행 속도를 높이기 위한 마지막 방법을 시도하거나, 반드시 배열을 사용하기 전에 각 항목에 의미 있는 값을 설정한다고 확신할 때 사용해야 함\
값을 초기화하지 않았기 때문에 시뮬레이션이나 게임에서 유용하게 사용할 수 있는 난수가 들어있다고 가정하면 안되는 것이, 이 숫자들은 통계적 이상 징후를 가지고 있어 무작위 표본 추출 시 빈약한 데이터를 만들게 됨
```py
numpy.empty(shape, dtype='float', order='C')
```
유일한 필수 인수인 shape 인수는 정수이거나 튜플이며 튜플은 더 높은 차원의 배열을 지정하는데 (3,3)은 2차원인 3x3 배열을 지정함\
dtype 인수는 각 항목의 데이터 타입을 정하며 기본값으로 'float'가 설정됨\
order 인수는 배열이 열-우선 혹은 행-우선 순서로 저장되는 것을 지정하는데 기본값은 'C'\
다음 예시는 16비트 signed 정수로 이루어진 2x2 배열을 생성함

````python
import numpy as np
a = np.empty((2,2), dtype='int16')
a
````

````python
import numpy as np
a = np.empty((2,2), dtype='float32')
a
````

### 5.10 fromfunction 함수
배열을 생성하는 가장 강력한 방법이며, fromfunction은 대상 배열의 색인을 인수로 사용하는 다른 함수를 사용하여 배열을 생성하고 초기화할 수 있음
```python
numpy.fromfunction(func, shape, dtype='float')
```
shape는 반드시 스칼라 값이 아닌 튜플을 사용해야 하는 규칙이 있기 때문에 1차원 데이터 세트를 생성하기 위해서는 (,5)와 같은 튜플 사용\
shape 튜플의 길이는 배열의 랭크(rank, 차원의 숫자)를 결정하며, 콜러블(callable)인 func에 필요한 인수의 숫자를 결정함\
5개의 자연수로 구성된 1차원 배열을 만드는 예시를 살펴보자

````python
import numpy as np

def simple(n):
    return n+1

a = np.fromfunction(simple, (5,), dtype='int32')
a
````

````python
a = np.fromfunction(lambda n:n+1, (5,), dtype=np.int32)
a
````

보통 일반적으로 사용되는 배열은 1차원보다 더 많은 차원을 갖는데, 다음 예시는 두 색인의 합계를 각 항목으로 초기화한 2차원 배열을 생성

````python
def add_it(r,c):
    return r+c
a = np.fromfunction(add_it, (3,3), dtype='int32')
a
````

````python
a= np.fromfunction(lambda r, c : r+c, (3,3), dtype=np.int32)
a
````

````python
# 각항목이 1의 값을 갖는 3x3 배열을 생성하려 햇지만 스칼라 1을 얻는 잘못된 방식 
a = np.fromfunction(lambda r, c : 1, (3,3), dtype='int32')
a
````

## 6. 불리언 배열:numpy에 마스킹

````python
import numpy as np
B = np.arange(1,10).reshape(3,3)
B
````

````python
B1 = B>4
B1
````

B1은 B와 같은 구조를 갖추었지만, 각 항목은 정수 대신 True 혹은 False이며 일반적인 규칙은 다음과 같음
* numpy 배열에 비교연산자(==, < 혹은 >)가 저기용될 때는 같은 구조의 불리언 배열을 반환함

이 배열을 사용하는 한 가지 방법은 4보다 크다는 조건을 만족하지 못하는 항목 모두에 0을 반영하고 나서, 두 배열(B와 B>4)을 곱함

````python
print(B*(B>4))
````

불리언 배열과 작업을 할 때, 비교 연산자는 낮은 우선순위를 갖기 때문에 괄호를 사용하는 것이 중요함\
불리언 배열을 더 좋게 사용하는 방법은 마스크로 사용하여 True인 항목과 같은 위치에 있는 항목은 선택하고, False 항목과 같은 위치에 있는 항목은 선택하지 않는 것임\
마스크로 불리언 배열을 사용하면 피연산자 배열의 구조와는 상관없이 1차원 배열을 생성함
```py
배열_이름[bool_array] # bool_array를 마스크로 사용하기
```
가령 7보다 큰 항목을 모두 선택하기 위해 마스크를 사용할 수 있음

````python
print(B[B>7])
````

````python
print(B[B % 3 == 1])
````

만약 and 혹은 or 키워드를 사용한 복잡한 조건이 필요하면 불리언과 동작해야 함에도 불구하고 다소 어려워짐\
좋은 해결책은 비트 연산자(&, |)를 불리언 마스크에 적용하는 것

````python
B2 = (B>2) & (B<7)
B2
````

````python
print(B[B2])
````

위 문법에서 대괄호 기호는 선택 항목이며, 반드시 넣을 필요는 없음\
일반적으로 단순히 plot을 호출할 때 2개의 1차원 배열인 x와 y를 인수로 넣음\
x를 생략하면 배열 [0, 1, 2, ..., N-1]이라고 가정하며 N은 y의 길이\
이렇게 하면 길이가 같은 x와 y에서 점들의 쌍을 구하여 그래프에 표시함\
x의 각 값은 y의 해당 갑소가 대응하여 (x,y)점을 얻으며 모든 점이 위치를 잡고 연결됨

````python
import numpy as np
import matplotlib.pyplot as plt

A = np.linspace(0, 2*np.pi)
plt.plot(A, np.sin(A))
plt.show()
````

linspace는 균일한 간격으로 총 N개의 값을 생성하는데, N의 기본값은 50\
따라서 배열 A에는 0부터 2*np.pi까지 두 값 사이에 균일한 간격으로 48개의 다른 부동소수점들이 채워짐

````python
# x와 y를 리스트 형식으로도 넣을 수 있음
plt.plot([0,1,2,3,4], [1,2,4,5,3])
plt.show()
````

````python
# plt.plot에서 x가 생략된 경우
plt.plot([1,2,4,5,3])
plt.show()
````

## 8. 여러 선 그래프 그리기
2개 이상의 그래프를 모두 포함하는 그래프를 그리기 위해서는 plot 함수를 여러 번 호출하면 됨

````python
A = np.linspace(0, 2*np.pi)
plt.plot(A, np.sin(A))
plt.plot(A, np.cos(A))
plt.show()
````

```py
plt.plot(X1, Y1, [fmt1,] X2, Y2, [fmt2,] ...)
```
x,y 쌍의 배열의 개수에는 제약이 없으며 선택적으로 포맷 인수를 추가할 수 있음

````python
plt.plot(A, np.sin(A), A, np.cos(A), 'o')
````

포매팅은 극적인 대비를 만들어 냄\
여기에 사인 곡선의 스타일을 지정하면 훨씬 더 뚜렷한 대비를 만들 수 있음\
다른 플로팅 함수인 title은 그래프의 제목을 지정하는 데 사용되며, xlabel과 ylabel은 각 축의 구체적인 라벨을 지정함

````python
plt.plot(A, np.sin(A), '^', A, np.cos(A), 'o')
plt.title('Sine and Cosine')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
````

## 9. matplotlib으로 히스토그램 만들기
히스토그램은 데이터를 보는 다른 방법을 제공함\
개별 데이터 점을 연결하는 대신 히스토그램은 데이터가 하위 범위에 포함되는 빈도를 보여줌\
데이터는 버킷(bucket) 혹은 통(bin) 안에 수집되며, 각 통은 범위를 의미함

````python
import numpy as np
import matplotlib.pyplot as plt

IQ_list = [91, 110, 105, 107, 135, 127, 92, 111, 105,
           106, 130, 145, 145, 128, 109, 108, 98, 129,
           100, 108, 114, 119, 99, 137, 142, 145, 112, 113]

IQ_A = np.array(IQ_list)

plt.hist(IQ_A)
plt.show()
````

완전한 문법을 확인해 보자
```py
plt.hist(A [, bins=10] [,keyword_args])
```
첫 번째 인수 A는 필수 인수이며, numpy 배열을 참조함\
bins 인수는 하위 범위, 혹은 통의 개수를 정의하며 기본 설정은 10, 가장 큰 값부터 가장 작은 값의 차이를 10으로 나누어서 하위 범위의 크기를 정함\
10이 아닌 다른 값으로 설정할 수도 있음

````python
plt.hist(IQ_list, bins=50)
plt.show()
````

````python
A = np.random.standard_normal(200000)
A = A * 10 + 100

plt.hist(A, bins=80, color='g')
plt.title('Normal Distribution in 80 bins')
plt.show()
````

## 11. 파이 차트 만들기
```py
plt.pie(array_data, labels=None, colors=None)
```
첫 번째 인수인 array_data는 각 카테고리의 상대적 크기를 지니고 있는 컬렉션\
labels 인수는 첫 번째 인수의 각 그룹에 해당하는 라벨을 문자열로 담고 있는 컬렉션\
colors 인수는 앞서 설명한 색을 지정하는 문자열을 담고 있는 컬렉션

````python
A_data = [3.7, 2.5, 1.9, 0.5]
A_labels = ['Poker', 'Chess', 'Comic Books', 'Exercise']
A_colors = ['k', 'g', 'r', 'c']
````

````python
plt.pie(A_data, labels=A_labels, colors=A_colors)
plt.title('Relative Hobbies of Dev. Team')
plt.axis('equal')
plt.show()
````
