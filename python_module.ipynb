# 파이썬 모듈
파이썬 프로그래밍이 익숙해지기 시작하면 코드를 여러 파일에 나누어서 작성하고 싶은 순간이 도래함\
이 여러 파일을 우리는 `모듈`이라고 부름\
단일 프로젝트에서 여러 모듈을 사용하면 많은 이점이 있음\
예를 들어 같은 프로젝트에 참여한 여러 개발자가 서로 다른 모듈에서 동시에 작업할 수 있음\
파이썬 모듈은 대부분 쉽게 사용할 수 있지만, 주의해야 할 사항들이 존재함

## 1. 파이썬 모듈의 개요
파이썬에서 모듈은 다른 것들과 마찬가지로 객체일 뿐이며, 자체 속성을 가짐\
이 사실로 알 수 있는 것은파이썬 스크립트는 최소한 하나의 모듈을 갖고 있어야 한다는 점\
`__name__`내장 식별자를 출력해보면 `__main__`이라는 이름의 모듈이 실행되고 있다는 것을 알 수 있음

````python
print(__name__)
````

이 장에서는 적어도 하나의 파이썬 소스 파일로 작업을 하고 있다고 가정함\
그리고 그 파일이 메인 모듈이 될 것이며 파일을 실행하면 실제 파일 이름이 무엇이든지 간에 이름이 `__main__`으로 변경됨

import 문을 사용하여 원하는 만큼의 다른 소스 파일을 가져온 후 패키지와 동일한 구문을 사용하여 프로젝트에 사용할 수 있음\
프로그램은 항상 메인 모듈을 실행하면 시작됨\
모듈을 탑재할 때는 일반적인 지침을 따라야 함

* import 문으로 다른 소스 파일을 탑재하는 것은 다른 패키지를 탑재하는 것과 같으며 .py 확장자는 포함하지 않음

* (함수의 로컬 변수가 아닌)모듈-수준 변수뿐만 아니라, 함수도 탑재할 수 있고, 정규화된 이름을 통해 가져온 변수도 참조할 수 있음\
예를 들어 e가 my_math.py 모듈에 정의된 경우 메인 모듈에서 my_math.e로 참조할 수 있음

* 한 가지 예외 사항은 변수 값을 절대로 변경하지 않는다면 직접 참조하는 것이 안전함\
가령 pi가 상수면 다른 모듈에서 간단하게 pi로 참조할 수 있음

* 모듈 간 서로 탑재하는 것은 피해야 함\
mod_a.py가 mod_b.py를 탑재하면 mod_b.py는 mod_a.py를 탑재하지 않아야 함\
또한 탑재의 순환적인 '루프'는 피해야 함\
만약 A가 B를 탑재하고 B가 C를 탑재한다면 C는 A를 탑재하지 않아야 함

순환 탑재 자체가 가능하지만 문제를 피하는 가장 간단한 방법은 순환 탑재를 하지 않는 것임

## 2. import 문의 변형
소스 파일이 다른 모듈에서 정의한 모듈-수준 심벌에 접근하는 방식으로 사용함
```py
import 모듈_이름
```

해당 모듈에서 생성한 함수는 적절한 지시자로 호출할 수 있음
```python
모듈_이름.함수_이름(args)
```

변수도 같은 방식으로 접근 가능함
```py
모듈_이름.변수
```

from 문을 사용하여 특정 심벌 이름을 나열하면 해당 심벌을 직접 사용할 수 있음
```py
from 모듈_이름 import 심벌1, 심벌2, ...
```

모듈의 모든 심벌을 탑재하도록 선택할 수도 있음
```py
from 모듈_이름 import *
```
