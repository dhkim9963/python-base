# 11. 피벗과 언피벗
## 11.1 피벗 테이블
원시 데이터는 여러 가지 열로 다양한 특성을 보유하고, 이러한 특성을 기반으로 원시 데이터를 다각도로 요약할 수 있음\
피벗 테이블(pivot table)은 원시 데이터를 다양한 속성으로 요약한 표이며, 판다스도 피벗 테이블을 생성함

### 11.1.1 피벗 테이블을 사용하는 이유
원시 데이터의 행은 개별 사례이며, 열은 특성(feature)\
원시 데이터는 데이터를 추가해 관리하기 쉬운 장점이 있지만 원시 데이터만으로는 데이터의 의미 파악이 어려움\
피벗 테이블은 원시 데이터를 특성별로 그룹화해서 집계한 테이블임\
원시 데이터와 피벗 테이블은 부족한 부분을 서로 채워주는 상호 보완의 관계이므로 반드시 병행하여 사용

### 11.1.2 피벗 테이블과 집계 함수
피벗 테이블은 원시 데이터를 특성으로 그룹을 나누어 각각의 셀에 요약함\
피벗 테이블은 원시 데이터를 그룹별로 하나의 셀에 요약하기에 차원을 축소하는 집계 함수가 필요함\
집계 함수는 차원 축소 함수인데, 평균, 표준편차 등의 함수뿐만 아니라 차원을 축소하는 함수는 모두 넓은 의미에서 집계 함수에 포함됨\
원시 데이터는 대부분 하나의 그룹에 많은 개수의 데이터가 있으며, 요약하려면 그룹의 값을 하나로 차원 축소하는 집계 함수가 필요함

### 11.1.3 피벗 테이블 생성하기
판다스 pivot_table\
원시 데이터로 피벗 테이블을 생성하는 함수
```py
pivot_table 함수의 주요 매개변수와 인수, 기본값
df.pivot_table(values=None, index=None, columns=None, aggfunc='mean')
```
* values : 집계할 대상 열을 지정함
* index : 행으로 구분할 그룹을 나누는 기준 열을 지정함
* columns : 열로 구분할 그룹을 나누는 기준 열을 지정함
* aggfunc : 그룹의 차원 축소할 집계 함수를 지정함

````python
import pandas as pd
# pd.options.display.float_format = '{:.1f}'.format # 소수점 첫째 자리 출력
data = [['김판다', '남', 'A', 97, '1등', '수시'],
        ['강승주', '여', 'A', 88, '2등', '정시'],
        ['권보아', '여', 'A', 78, '3등', '정시'],
        ['조민영', '여', 'B', 64, '3등', '수시'],
        ['박상현', '남', 'B', 84, '2등', '수시'],
        ['송중기', '남', 'B', 89, '1등', '수시'],
        ['최진환', '남', 'C', 87, '3등', '정시'],
        ['장범준', '남', 'C', 92, '2등', '정시'],
        ['안지선', '여', 'C', 99, '1등', '수시']]
col = ['이름', '성별', '반', '점수', '반등수', '비고']

data1 = {'제품': ['A', 'B', 'A', 'B', 'B', 'A'],
         '판매량': [float('nan'), 2, 3, 4, float('nan'), 6]}

df = pd.DataFrame(data, columns=col)
df1 = pd.DataFrame(data1)
df
````

df를 반과 성별로 나누어 평균 점수를 피벗 테이블로 만들자\
pivot_table 함수의 주요 매개변수
1. values : 집계 대상
2. index : 행으로 구분할 그룹
3. columns : 열로 구분할 그룹
4. aggfunc : 집계 함수

values에는 집계할 대상인 열을 입력해야 함\
df의 반과 성별로 나누어 평균 점수를 구할 것이기에 집계 대상은 점수 열

그룹 구분 기준이 되는 열을 index와 ㅊolumns에 지정해야 함\
피벗 테이블의 행에서 반을 구분하고 열에서 남녀를 구분할 것이기에 index에 반, columns에 성별을 입력함

aggfunc은 반과 성별로 나누어 평균 점수를 피벗 테이블로 생성할 것이므로 집계함수에 mean을 입력함

````python
df.pivot_table(values='점수', index='반', columns='성별', aggfunc='mean')
````

### 11.1.4 복수의 인수를 입력해 피벗 테이블 생성하기
pivot_table 함수의 네 가지 매개변수는 모두 복수의 인수를 입력할 수 있으며 입력할 인수를 리스트로 묶어 입력함\
index에 '반'과 '성별'을 리스트로 묶어 입력하고, columns에 '비고'를 지정

````python
df.pivot_table(values='점수', index=['반','성별'], columns='비고')
````

매개변수 index에 복수의 열을 입력해 피벗 테이블로 멀티 인덱스를 보유한 데이터 프레임을 반환함\
그룹에 해당하는 데이터가 존재하지 않는 셀은 NaN을 반환함\
이번에는 복수의 집계 함수를 사용하자

````python
df.pivot_table('점수',index='반', columns='성별', aggfunc=['mean','count'])
````

복수의 인수를 입력하면 멀티 인덱스를 보유한 데이터 프레임을 반환할 때가 많으니 멀티 인덱스를 다뤄야 함

### 11.1.5 인수로 함수를 입력하는 방법
pivot_table 함수는 차원 축소를 하고자 집계 함수를 입력해야 하기에 함수를 인수로 입력하는 방법에 대해 살펴볼 것임\
판다스에서는 pivot_table 함수뿐 아니라 apply 함수 등 인수로 함수를 입력해야 할 때가 자주 발생하는데 이 방법으로 함수를 입력함
1. 대상에 적용되는 함수를 입력함\
pivot_table 함수에 인수로 입력하는 집계 함수는 시리즈가 적용 대상인 함수여야 함\
시리즈에 적용되는 함수라면 사용자 정의 함수와 lambda 함수도 입력할 수 있음\
시리즈에 apply를 사용할 때는 셀에 적용되는 함수, 데이터 프레임에 apply를 사용할 때는 시리즈에 적용되는 함수를 인수로 입력해야 함
2. 메서드가 아닌 함수 형태로 입력함\
함수를 인수로 입력하기에 메서드가 아닌 함수를 입력해야 함\
null이 아닌 값의 개수를 세는 함수는 pd.Series.count인데 시리즈에 적용할 때는 s.count()의 형태로 메서드에 사용하지만 인수로 입력할 때는 함수 형태인 pd.Series.count로 입력해야 함
3. 자주 쓰이는 함수는 문자열로 해당 함수를 호출할 수 있으니 문자열을 입력함\
'count', 'sum', 'max', 'first', 'last' 등ㅇ의 많은 함수를 문자열로 입력함

### 11.1.6 그룹화에만 적용되는 집계 함수(first, last)
일반적으로 pivot_table에 쓰이는 집계 함수는 일반적인 통계 함수이며 대부분 시리즈에도 적용이 가능함\
하지만 first 함수, last 함수는 pivot_table 함수 혹은 groupby 함수처럼 그룹을 나누는 함수와 함께 사용할 때만 사용이 가능함

````python
df1
````

````python
df1.pivot_table('판매량', index='제품', aggfunc='first')
````

````python
df1.pivot_table('판매량', index='제품', aggfunc='last')
````

````python
(df.sort_values('점수', ascending=False)
 .pivot_table('이름', index='반', columns='성별', aggfunc='first')
)
````

### 11.1.7 문자열 피벗
판다스 pivot\
집계 함수 없이 피벗 테이블을 만드는 함수로 주로 문자열 피벗에 쓰임
```py
pivot 함수의 주요 매개변수와 인수, 기본값
df.pivot(index=None, columns=None, values=None)
```
pivot_table 함수에 first, last와 같은 특정 집계 함수를 지정해 문자열 피벗을 수행하기도 하나, 통상 pivot 함수를 사용해 문자열의 피벗을 수행함\
pivot_table 함수에서 집계 함수가 필요했던 이유는 여러 개의 데이터가 하나의 그룹에 소속되어 차원을 축소해야 했기 때문\
모든 그룹에 각각 하나의 데이터만 소속된다면 집계 함수가 필요하지 않으며 이때 사용하는 함수가 pivot 함수\
집계 함수 없이 사용하기에 반드시 하나의 그룹에는 둘 이상의 데이터가 소속되지 않아야 하며, 만약 소속된 데이터가 2개 이상인 그룹이 존재하면 ValueError가 발생함

````python
df.pivot(index='반', columns='반등수', values='이름')
````

### 11.1.8 빈도수를 집계하는 교차표 생성하기(crosstab)
판다스 crosstab\
빈도수를 집계해 교차표를 생성하는 함수

````python
data = {'반': ['A', 'A', 'A', 'B', 'B', 'B'],
        '성별': ['남', '여', '여', '여', '남', '남']}
df = pd.DataFrame(data)
df
````

기본적으로 빈도수는 value_counts 함수로 집계하며, pivot_table 함수의 집계 함수로 count 함수를 사용하여 집계하기도 함\
value_counts는 교차표가 아닌 시리즈를 생성하며, pivot_table 함수로 빈도수를 집계해 교차표를 생성하려면 최소 세 개의 열을 지정\
따라서 위처럼 열이 2개뿐일 때 pivot_table 함수를 사용하여 빈도수를 파악하려면 의미 없는 집계용 열을 추가해야 함\
이때 사용하는 함수가 crosstab 함수이며, crosstab 함수는 메서드로 사용할 수 없고 반드시 함수 pd.crosstab 형태로 사용해야 함

````python
pd.crosstab(df['반'],df['성별'])
````

## 11.2 언피벗
언피벗은 피벗의 반대이며 언피벗을 수행하면 원시 데이터처럼 개별 사례인 행과 특성인 열을 가진 데이터 구조로 변환됨

### 11.2.1 언피벗이 필요한 이유
원시 데이터는 데이터 추가 및 관리에 편리하고 피벗 테이블은 데이터 의미 파악에 용이함\
각자 장점이 있으므로 역할에 맞게 활용하고 서로 보완해야 하지만 실무 데이터는 그렇지 않을 때가 많음\
원시 데이터는 무겁고 다루기 어려워 피벗 테이블만 남겨서 사실상 피벗 테이블로 데이터를 관리할 때가 많음\
하지만 피벗 테이블에서 새로운 피벗 테이블을 만드는 것은 쉽지 않기에 원시 데이터도 필요함\
원시 데이터가 있다면 새로운 피벗 테이블을 쉽게 만들 수 있지만 원시 데이터는 무겁고 다루기 힘들기에 필요한 것이 언피벗

언피벗은 피벗의 반대이며, 피벗 테이블을 원시 데이터와 유사한 구조의 데이터로 만드는 작업\
원시 데이터가 보존되지 않은 때는 언피벗으로 피벗 테이블을 원시 데이터와 유사한 데이터로 변환함\
언피벗을 수행하면 원시 데이터의 장점을 그대로 계승하기에 데이터의 추가도 수월함\
언피벗을 하면 원시 데이터와 같은 형태의 구조가 되어 행 추가만 해도 손쉽게 개별 사례들을 추가하고, 새로운 열을 추가하는 것으로 새로운 특성을 생성함

판다스는 크게 두 가지 방법으로 언피벗을 수행함\
첫째, stack 함수와 reset_index 함수를 조합하여 언피벗을 수행함\
둘째, melt 함수로 언피벗을 수행함

### 11.2.2 stack 함수로 언피벗

````python
import pandas as pd
data1 = {'1등': {'1회차': '김판다', '2회차': '박효신', '3회차': '김판다'},
         '2등': {'1회차': '권보아', '2회차': '권보아', '3회차': '박효신'},
         '3등': {'1회차': '박효신', '2회차': '강승주', '3회차': '김범수'}}
df = pd.DataFrame(data1)
df
````

stack은 앞서 다룬 멀티 인덱스에 활용하는 함수로 언피벗을 위한 함수는 아니지만 언피벗에 활용함\
언피벗된 상태는 데이터가 하나의 열로 늘어선 상태인데 ㄴtack을 해도 시리즈가 되면서 데이터가 한 줄로 늘어섬\
이때 reset_index로 인덱스를 모두 열로 만들면 언피벗된 데이터 프레임이 됨

````python
df.stack()
````

````python
df.stack().reset_index()
````

````python
df1 = df.stack().reset_index().set_axis(['회차','등수','이름'], axis=1)
df1
````

언피벗을 수행하였으니 df1으로 새로운 피벗 테이블을 만들 것인데, 각 사람의 등수의 빈도수를 집계할 것임\
pivot_table 함수의 count도 가능하지만 빈도수를 집계하는 것이기에 crosstab 함수를 사용하는 것이 더 간편함

````python
pd.crosstab(df1['이름'],df1['등수'])
````

### 11.2.3 melt 함수로 언피벗
판다스 melt\
데이터 프레임의 언피벗을 수행하는 함수
```py
df.melt(id_vars=None, value_vars=None, var_name='variable', value_name='value')
```
* id_vars : 인덱스처럼 언피벗하지 않고 별도의 열로 처리될 열을 지정함
* value_vars : 언피벗을 수행할 열을 지정하며 기본값으로 id_vars에 지정되지 않은 모든 열을 언피벗함
* var_name : 언피벗 후 생성되는 variable 열의 이름을 변경함
* value_name : 언피벗 후 생성되는 value 열의 이름을 변경함

````python
import pandas as pd
data = {'반': ['A', 'B', 'C'], '남': [10, 20, 15], '여': [30, 40, 35]}
df = pd.DataFrame(data)
df1 = df.set_index('반')
df
````

df의 반 열은 언피벗 대상이 아니기에 인덱스로 설정되어야 함\
하지만 반 열이 일반 열로 설정되어 있기에 stack 함수로 df의 언피벗을 수행하려면 반 열을 인덱스로 설정하고 언피벗을 수행해야 함

melt 함수는 stack 함수로 언피벗할 때의 인덱스처럼 언피벗을 수행하지 않을 열을 지정할 수 있음\
매개변수 id_vars에 '반'을 입력하면 반 열의 데이터는 제외하고 나머지 데이터가 언피벗됨\
언피벗을 수행할 열은 value_vars에 지정하는데 남,여 열 모두 언피벗을 수행

````python
df.melt('반', value_vars=['남','여'])
````

````python
df.melt('반')
````

````python
df.melt('반', var_name='성별', value_name='인원수')
````

### 11.2.4 stack 함수와 melt 함수의 언피벗 수행의 차이점
melt 함수는 인덱스를 보존하지 않아 통상은 인덱스에 필요한 데이터가 존재한다면 stack 함수를 사용함\
두 함수의 언피벗 결과는 정렬 순서가 다르지만 언피벗의 목적은 새로운 피벗 테이블 생성이기에 정렬 상태의 의미는 아주 크지 않음\
stack 함수와 melt 함수 모두 언피벗을 수행하므로 본인에게 편한 함수로 언피벗을 시도하면 됨

```py
groupby 함수의 주요 매개변수와 인수, 기본값
df.groupby(by=None, level=None, as_index=True, sort=True, group_keys=True)
```

* by : 그룹을 나누는 기준을 지정하는데 주로 열을 지정하며 복수라면 리스트로 입력함

* as_index : groupby 함수로 집계할 때 그룹이 인덱스가 될지 지정함
* sort : 그룹의 키로 정렬할지 지정함
* group_keys : groupby와 apply를 함께 사용할 때, 그룹의 키를 인덱스에 추가할지 지정함

````python
import pandas as pd
data1 = {'반': ['A', 'B', 'A', 'B', 'B', 'A'],
         '점수': [88, 30, 78, 99, 62, 85]}
df = pd.DataFrame(data1)
df
````

### 12.1.4 groupby 함수로 행 간의 연산하기

````python
import pandas as pd
# pd.options.display.max_rows = 8 # 8행까지만 출력
data = [['2025-01-03', '판다전자', 10000],
        ['2025-01-03', '성심당', 2000],
        ['2025-01-04', '성심당', 1600],
        ['2025-01-04', '판다전자', 12000],
        ['2025-01-05', '판다전자', 15000],
        ['2025-01-05', '성심당', 2000],
        ['2025-01-06', '판다전자', 13500],
        ['2025-01-06', '성심당', 2400]]

df = pd.DataFrame(data, columns=['날짜', '종목', '종가'])
df
````

df에 전일 종가를 새로운 열로 만들고자 하면 shift 함수를 이용하여 시리즈를 지정한만큼 밀 수 있음\
단일 종목만 있었다면 전일 종가 데이터를 쉽게 열로 만들었을 것이지만 df는 판다전자와 성심당의 데이터가 섞여 있음\

````python
df['전일종가'] = df.groupby('종목')['종가'].shift()
df
````

shift와 비슷한 함수 diff와 pct_change 함수도 groupby와 사용하는데 이번에는 전일 대비 등락 열과 등락률 열을 생성

````python
df['등락'] = df.groupby('종목')['종가'].diff()
df
````

````python
df['등락률'] = df.groupby('종목')['종가'].pct_change()
df
````

````python
import pandas as pd
data = [['2021-01-01', '김판다', 10000], ['2021-01-01', '강승주', 2000],
        ['2021-01-02', '김판다', 20000], ['2021-01-02', '강승주', 5000],
        ['2021-01-03', '강승주', 8000], ['2021-01-03', '김판다', 5000]]
df = pd.DataFrame(data, columns=['날짜', '이름', '입금액'])
df
````

### 12.1.9 transform 함수로 사용자 저으이 함수 적용하기

열 가공 함수이지만 메서드로 활용할 수 없다면 transform 함수가 좋은 방법임\
대부분의 열 가공 함수가 메서드로 활용할 수 있으나, cut 함수나 qcut 함수처럼 메서드로 활용할 수 없는 함수들이 일부 존재함\
또한 사용자 정의 함수는 태생적으로 메서드로 활용할 수 없으니 transform 함수가 필요함

````python
data1 = {'상품종류': ['스낵', '스낵', '스낵', '핸드폰', '핸드폰', '핸드폰'],
        '상품코드': ['S001', 'S002', 'S003', 'P001', 'P002', 'P003'],
        '가격': [6000, 10000, 4000, 10000, 1000000, 500000]}
df1 = pd.DataFrame(data1)
df1
````

df1은 각 상품의 카테고리 분류와 가격에 대한 데이터인데 각 상품을 가격에 따라 저가, 중가, 고가로 균등 분할해 또 다른 범주화를 수행\
전체를 1/3로 나누어 저가, 중가, 고가로 분류하되 같은 상품 종류 내에서 분류할 때 qcut 함수를 활용하면 손쉽게 범주화가 가능함\
하지만 qcut 함수는 반드시 pd.qcut인 함수 형태로 사용해야 하기에 transform 함수가 필요하며 lambda를 사용하면 편리함

````python
df1['분류'] = df1.groupby('상품종류')['가격'].transform(
    lambda x : pd.qcut(x, q=3, labels=['저가','중가','고가'])
)
df1
````

````python
# lambda를 사용하지 않고 진행하는 방법
df1['분류'] = df1.groupby('상품종류')['가격'].transform(
    pd.qcut, q=3, labels=['저가','중가','고가']
)
df1
````

### 12.2.2 groupby 함수로 집계하기와 피벗 테이블의 차이
groupby 함수를 사용하면 그룹 집계를 수행할 수 있으며 pivot_table 함수 역시 그룹 집계를 수행함\
그렇다면 groupby 함수의 그룹 집계와 피벗 테이블은 어떤 차이가 존재하는가?\
결론적으로 두 방식은 결과의 구조만 다를 뿐 비슷한 기능을 수행함

````python
import pandas as pd
# pd.options.display.float_format = '{:.2f}'.format # 소수점 둘째 자리 출력
data1 = [['김판다', 'A', '남', 95, 90], ['최진환', 'B', '남', 93, 90],
         ['조민영', 'B', '여', 88, 80], ['박연준', 'A', '남', 85, 70],
         ['강승주', 'B', '여', 78, 90], ['안지선', 'A', '여', 72, 70]]
df1 = pd.DataFrame(data1, columns=['이름', '반', '성별', '국어', '영어'])
df1
````

````python
df1.groupby(['반', '성별'])['국어'].mean()
````

````python
df1.pivot_table('국어', index='반', columns='성별', aggfunc='mean')
````

````python
df1.groupby(['반', '성별'])['국어'].mean().unstack()
````

````python
df1.pivot_table('국어', index=['반', '성별'], aggfunc='mean')
````

동일한 집계 결과를 반환할 수 있지만 간단한 교차표를 만들 때는 pivot_table 함수가 더 편리함\
복잡한 구조의 피벗 테이블은 오히려 groupby 함수가 뛰어난데, 예를 들어 복수의 열을 집계할 때는 groupby 함수가 더 편리함\
groupby 함수로 복수의 열을 한 번에 집계할 때는 복수의 열을 리스트로 묶어 대괄호에 입력함(pivot_table 함수로도 복수의 열을 집계하는 것이 가능하지만 groupby 쪽이 더 편리함)

````python
df1.groupby(['반', '성별'])[['국어','영어']].mean()
````

### 12.2.3 groupby 함수와 agg 함수로 집계하기
groupby 함수는 agg 함수와 조합해 맞춤형 피벗 테이블 생성이 용이하며 agg 함수는 맞춤형 집계에 큰 강점을 가짐\
agg 함수는 aggregate의 약자로 집계 함수를 적용하는 것에 특화된 함수이며 여러 가지 간편한 부가 기능을 제공함\
여러 집계 함수를 동시에 적용할 수 있고 열마다 다른 집계 함수를 사용할 수 있음

agg 함수의 장점
1. 여러 집계 함수를 동시에 적용할 수 있음\
$\rightarrow$ 복수의 집계 함수를 리스트로 묶어 인수로 입력
2. 열마다 다른 집계 함수를 사용할 수 있음\
$\rightarrow$ 적용할 열과 집계 함수를 딕셔너리의 키와 밸류로 묶어 인수로 입력
3. 열마다 다른 집계 함수를 사용하고, 집계된 열 이름을 지정할 수 있음\
$\rightarrow$ 열 이름을 매개변수로 입력하고, 집계 함수를 적용할 열과 함수를 튜플로 묶어 인수로 입력

````python
df1
````

df1에서 반과 성별을 기준으로 그룹을 나누어 국어의 최고 점수와 최저 점수를 동시에 구하자\
반과 성별을 리스트로 묶어 by에 입력하고, 국어 열을 대괄호로 인덱싱한 후 agg 함수를 적용\
agg 함수의 인수로 최댓값을 구하는 'max'와 최솟값을 구하는 'min'을 리스트로 묶어 입력함

````python
df1.groupby(['반','성별'])['국어'].agg(['max','min'])
````

열마다 다른 집계 함수도 적용해 보자\
영어 열은 평균을 집계하고, 국어 열은 인원수를 집계하자\
적용할 열과 집계 함수를 딕셔너리의 키와 밸류로 묶어 agg 함수의 인수로 입력하며, agg 함수에서 함수를 적용할 열과 함수를 개별적으로 지정하므로 대괄호 인덱싱은 생략함

````python
df1.groupby(['반', '성별']).agg({'영어' : 'mean', '국어' : 'count'})
````

위 예제의 결과를 보면 집계 결과의 열 이름은 집계 대상 열과 동일한데 집계 결과의 열 이름을 수정해야 할 때도 존재함
```py
df1.groupby().agg(새로운_열_이름 = ('대상열_이름', 함수))
```
위와 같은 형태로 사용함

````python
df1.groupby(['반', '성별']).agg(
    영어평균 = ('영어', 'mean'),
    국어평균 = ('국어', 'mean'),
    인원수 = ('국어', 'count')
)
````

### 12.2.4 문자열의 결합
파이썬에는 문자열을 결합하는 join 함수가 있는데, 구분자를 지정하고 결합 대상인 배열을 입력하면 배열의 문자열을 구분자로 구분해 하나로 결합해주는 함수로, 리스트뿐 아니라 시리즈에도 사용함

````python
s = pd.Series(['A', 'BC', 'DEF'])
'/'.join(s)
````

시리즈에 적용 가능한 함수이지만 단독으로 groupby 함수에 적용할 수는 없음\
groupby 함수와 조합하는 함수는 메서드 형태여야 하는데, join 함수는 판다스 함수가 아니므로 시리즈에 메서드로 적용할 수 없음\
이처럼 메서드로 사용할 수 없는 집계 함수는 agg 함수와 조합해 사용함

````python
df1
````

````python
df1.groupby('영어').agg(학생=('이름', '/'.join))
````

### 12.2.5 agg 함수로 사용자 정의 함수 적용하기
사용자 정의 함수 역시 메서드 형태로 사용할 수 없기에 agg 함수에 입력해야 함\
이때 적용할 사용자 정의 함수는 반드시 차원이 축소되는 집계 함수여야 함\
생성한 사용자 정의 함수가 집계 함수가 아니라면 transform 함수를 사용함\
국어 열의 최대와 최소의 격차를 lambda 함수로 집계할 것인데 최대와 최소의 격차는 차원이 축소되는 집계 함수이기에 agg 함수로 적용

````python
df1.groupby('반').agg(
    최대_최소=('국어', lambda x: x.max() - x.min())
)
````

최대, 최소를 구하여 뺄셈을 해도 격차를 구할 수 있지만, lambda 함수를 사용해 한번에 집계한다면 agg 함수에 lambda 함수를 입력함\
조건에 맞는 대상만 집계할 때도 lambda 함수가 유용한데 where 함수를 사용하면 조건에 맞지 않는 대상을 NaN으로 변환함\\
mean처럼 판다스의 집계 함수는 NaN을 제외하고 집계를 수행하기에 where 함수와 mean 함수를 조합하면 조건에 맞는 대상의 평균만을 구함

````python
cond = df1['성별'] == '남'
df1['국어'].where(cond).mean()
````

````python
cond = df1['성별'] == '남'
df1.groupby('반').agg(
    국어_남자_평균=('국어', lambda x: x.where(cond).mean()),
    국어_여자_평균=('국어', lambda x: x.where(~cond).mean()),
    국어_전체_평균=('국어', 'mean')
)
````

### 12.2.6 agg 함수와 transform 함수의 차이
지금까지 함수를 적용하는 함수인 agg 함수와 transform 함수를 학습하였는데 두 함수의 차이점은 다음과 같음
* agg 함수 : 집계할 때 사용하며, 집계 함수만 입력이 가능하고, 집계 결과를 반환함
* transform 함수 : 열을 생성할 때 사용하며 열을 만들 수 있는 시리즈를 반환함\
시리즈를 생성하는 다양한 함수를 입력할 수 있으며, 집계 함수를 입력해도 시리즈를 반환함

````python
df
````

````python
df.groupby('반')['점수'].agg('mean')
````

````python
df.groupby('반')['점수'].transform('mean')
````

## 12.3 groupby 심화
### 12.3.1 groupby 함수의 여러 가지 매개변수

````python
import pandas as pd
data1 = [[1, 'A', '반장', '물리001', 94],
         [1, 'A', '부반장', '화학001', 90],
         [2, 'A', '부반장', '화학002', 85],
         [2, 'A', '반원', '물리002', 95],
         [1, 'B', '부반장', '물리003', 86],
         [1, 'B', '반장', '화학003', 75],
         [1, 'B', '반원', '화학004', 80]]
df = pd.DataFrame(data1, columns=['학년', '반', '직책', '학생코드', '점수'])
df1 = df.set_index(['학년', '반']).copy()
df
````

df는 학생마다 학년, 반, 반에서의 직책, 점수가 나타나고, 학생코드 열에는 선택 과목이 표현된 코드가 입력됨\
df1은 df와 유사한 데이터이지만, 학년과 반의 데이터가 인덱스에 위치함

as_index는 집계 후 그룹을 인덱스에 위치시킬지, 일반 열로 만들지를 결정하는 매개변수\
기본값은 True라 이제까지의 집계 결과는 각 그룹이 인덱스에 위치하였지만 False를 입력하면 일반 열로 생성됨

````python
# groupby 함수의 매개변수 as_index 활용
df.groupby(['학년', '반'], as_index=False)['점수'].mean()
````

각 그룹이 인덱스가 아닌 일반 열이 되므로 집계 결과도 시리즈가 아니라 데이터 프레임으로 반환됨\
as_index=False를 사용하지 않고 집계 결과에 추가로 reset_index 함수를 적용해도 동일한 결과를 얻음

df의 데이터는 2학년 A반이 1학년 B반보다 먼저 등장하지만 위의 결과를 보면 1학년 B반이 2학년 A반보다 우선함\
그룹의 정렬을 수행한 후 집계하기 때문인데 sort=False를 입력하면 정렬을 수행하지 않음

````python
df.groupby(['학년','반'], sort=False)['점수'].mean()
````

df1처럼 그룹의 기준이 인덱스에 위치한다면 by 대신 level로 그룹을 나누는 기준을 지정함

````python
df1
````

````python
df1.groupby(level=[0,1])['점수'].mean()
````

### 12.3.4 그루퍼
groupby 함수는 열뿐만 아니라 인덱스를 기준으로도 그룹을 나누고, 입력한 배열을 기준으로도 그룹을 나눔\
그룹의 기준을 데이터 프레임 내부에서 지정하는 것이 아니라 외부의 배열로 지정할 때, 그룹의 기준이 되는 배열을 그루퍼라고 함\
df의 학생코드 열은 각 학생의 선택 과목을 나타냄\
선택 과목을 기준으로 집계를 수행하자\
학생 코드 열에서 선택 과목의 데이터를 추출해 별도의 열을 생성하면 groupby 함수가 적용됨\
이때 별도의 열을 생성하지 않고 가장 위 행부터 차례로 그룹을 지정해 리스트로 입력해도 그룹을 나눌 수 있음

````python
df
````

````python
grp1 = ['물리', '화학', '화학', '물리', '물리', '화학', '화학']
df.groupby(grp1)['점수'].mean()
````

변수 grp1이 그룹을 나누는 기준이 되는 배열이고 이를 그루퍼라고 지칭함\
다만 리스트로 직접 생성하는 방식은 지나치게 번거로운데 7행 정도라면 가능하지만 빅데이터에서는 불가능함\
주로 원본 데이터를 처리한 결과를 그루퍼로 활용함

df1의 학생코드 열에서 각 셀의 앞 두글자만 인덱싱하면 과목의 데이터가 되는데 이 시리즈를 그루퍼로 사용

````python
grp2 = df['학생코드'].str[:2]
df.groupby(grp2)['점수'].mean()
````

그루퍼를 활용하면 별도의 열을 생성하지 않고 기존 데이터를 가공하는 것만으로 그룹 집계가 가능함\
그루퍼는 원본 데이터를 변환하지 않는 장점이 있으며, 특히 열이 없는 시리즈에 groupby 함수를 적용할 때는 대부분 그루퍼를 사용함

### 12.3.5 transform 함수를 활용한 다양한 그룹 필터링(all, any)
groupby 함수와 transform 함수를 활용하여 그룹 필터링을 수행할 때 특정 데이터의 존재 여부로 그룹을 필터링해볼 것임

df를 복제한 df2에서 반장이 존재하는 데이터만 추출하고, 반장 여부를 True, False로 반환하는 반장 열을 생성\
그리고 학년과 반으로 그룹을 나누어 반장 열에 transform 함수를 적용함\
transform 함수에 집계 함수 any를 입력하면 True가 존재하는 그룹에만 True인 시리즈를 싱성하기에 반장이 존재하는 반을 필터링함

````python
df2 = df.copy()
df2
````

````python
df2['반장'] = df2['직책'].eq('반장')
cond1 = df2.groupby(['학년', '반'])['반장'].transform('any')
df2[cond1]
````

any 함수와 all 함수를 transform 함수와 조합해 특정 데이터를 보유한 그룹 전체를 필터링하는 것은 유용한 기법\
그루퍼를 사용하면 별도의 열을 생성하지 않고도 깔끔하게 같은 작업을 수행함

````python
cond2 = df['직책'].eq('반장').groupby([df['학년'], df['반']]).transform('any')
df[cond2]
````

# 13. 시계열 데이터
# 13.1 시계열 데이터 소개
시계열 데이터는 시간의 흐름에 따라 수집된 데이터를 의미함\
대부분의 데이터는 시간의 흐름에 따라 발생하기에 시계열 데이터를 다루는 기능을 익히는 것은 중요함\
판다스는 금융 데이터를 다루고자 만들어진 라이브러리이기에 시계열 데이터 처리에 특화되어 있음\
이번 장에서는 시계열 데이터로의 변환, 생성 그리고 시계열에서 가장 중요한 함수인 시계열 데이터 기반의 그룹화 함수 resample 학습

### 13.1.1 시계열 데이터의 종류
판다스에서 주로 다루는 시계열 데이터는 timestamp, period, timedelta 세 종류가 있음

timestamp는 특정 시점을 가리키는 시계열 데이터\
기본적으로 날짜와 시간이 모두 표기되며 표기 가능한 최소 시간 단위는 ns(nano second)

period는 timestamp 사이의 기간을 나타내는 시계열 데이터\
타임 스탬프가 1달 간격이라면 period는 2022년 1월을 뜻하는 2022-01로 표현되며 해당 기간을 전부 포함한다는 의미

timedelta는 타임스탬프 사이의 시간 차이를 나타내는 시계열 데이터\
두 시점의 정확한 시간 차이는 31 days 00:00:00이며 이러한 시간 차이를 timedelta라고 함

세 종류의 시계열 데이터는 모두 각자의 쓰임이 있으며 세 자료형에 모두 숙달되면 편리함\
period와 timedelta는 다른 자료형으로 대체해서 다룰 수 있기에 시계열에서 가장 중요한 학습 목표는 timestamp 데이터를 다루는 것

### 13.1.2 datetime 자료형
timestamp와 datetime의 관계\
1640962800 $\rightarrow$ 정수 형식의 timestamp\
2022-01-01 00:00:00 $\rightarrow$ 날짜와 형식의 timestamp (datetime)\
timestamp를 날짜와 시간 형식으로 표현한 자료형이 datetime 자료형

datetime의 출력\
원칙적으로 datetime은 timestamp(시점)를 나타내기에 날짜와 시간이 모두 출력됨\
ex) 2022-01-01 00:10:01\
다만 모든 데이터가 00:00:00이라면 시간을 생략하고 날짜만 출력\
ex) 2022-01-01 00:00:00 $\rightarrow$ 2022-01-01

### 13.1.3 시계열 데이터의 학습 주안점
1. 변환(to_datetime)
2. 주기
3. 생성(date_range)
4. 그룹화와 그룹 집계(resample)
5. 특정 시계열 데이터 추출(dt 접근자와 다양한 메서드)

먼저 시계열 데이터로 변환하는 방법을 배우는데 특히 datetime 자료형으로 변환해야 하며 시리즈는 to_datetime 함수로 변환함\
파일에서 데이터 프레임으로 불러올 때 datetime 자료형으로 미리 지정할 수도 있는데 이때는 매개변수 parse_dates를 사용함

다음으로 시계열 데이터의 생성과 그룹화에 쓰일 주기를 알아야 하며 학습된 주기를 기반으로 시계열 데이터를 생성하는 것도 학습함

주기를 기반으로 일명 시계열의 groupby 함수라 불리는 resample 함수를 사용해 시계열 데이터를 손쉽게 그룹화하는 방법도 학습함\
resample 함수를 숙지하는 것이 시계열 단원에서 초반 목표

마지막으로 시계열 데이터에서 특정 데이터만 추출하고자 dt 접근자와 함께 사용되는 다양한 속성과 메서드를 활용\
특히 dt.strftime 함수로 datetime 자료형을 원하는 형태의 문자열로도 바꿀 수 있음

## 13.2 시계열 데이터로 변환 및 인덱싱
### 13.2.1 시계열 데이터로 변환하기(to_datetime)
판다스 to_datetime\
datetime 자료형으로 변환하는 함수
```py
to_datetime 함수의 주요 매개변수와 인수, 기본값
pd.to_datetime(arg, errors='raise', utc=False, format=None, origin='unix')
```
* arg : datetime으로 변환할 데이터를 입력하며, 단일 값이거나 1차원 배열만 가능함
* errors : 변환 불가능한 데이터를 처리하는 방법을 지정함
* utc : 시간대(timezone)가 포함된 데이터를 협정 세계시(UTC)로 변환할지 지정함
* format : 문자열의 날짜 표기 형식을 지정함
* unit : timestamp가 숫자로 주어질 때 숫자의 기본 단위를 지정하며, 기본값은 'ns'
* origin : timestamp가 숫자로 주어질 때 기준 날짜를 지정함

````python
import pandas as pd
data = ['2023-01-01', '2023-02-02', '2023-03-02', '2023-04-10', '2023-05-31']
s = pd.Series(data)
s1 = pd.Series(['2022-01-03', '김판다'])
s
````

시리즈 s1을 datetime 자료형으로 변환하는데 to_datetime 함수는 메서드로 사용할 수 없고 함수 pd.to_datetime 형태로 사용함

````python
pd.to_datetime(s)
````

to_datetime 함수는 단일 값 또는 1차원 배열만 인수로 입력되어 시리즈만 변환함\
데이터 프레임에서 여러 열의 시계열 데이터를 가질 때가 흔하지 않고, apply 함수를 이용하면 데이터 프레임 전체에 적용할 수 있기 때문

to_datetime 함수에 errors='coerce'를 입력하면 변환 불가능한 데이터가 섞여 있을 때 NaT로 변환하고, 나머지 데이터를 datetime으로 변환하는데 NaT(Not a Time)는 시계열 데이터에서 사용되는 null

````python
print(s1, '\n') # s1과 빈 행을 출력하는 코드
pd.to_datetime(s1, errors='coerce')
````
